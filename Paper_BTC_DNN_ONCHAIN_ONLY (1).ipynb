```python
# ============================================================
# BITCOIN ON-CHAIN FORECASTING MODEL — ONCHAIN ONLY
# LSTM / GRU / BI-LSTM Ensemble — Ablation Study (On-Chain Features Only)
# ============================================================
# Struttura:
#   1.  Setup & Import
#   2.  Caricamento dati on-chain
#   3.  Normalizzazione on-chain
#   4.  Configurazione flags & parametri
#   5.  Preprocessing & Split temporale
#   5b. Walk-Forward Validation (con gap anti-leakage)
#   6.  Hyperparameter Tuning (Optuna)
#   7.  Modello LSTM
#   8.  Modello GRU
#   9.  Modello BI-LSTM
#   10. Valutazione individuale + Feature Importance
#   11. Ensemble adattivo con filtro qualità
#   12. Test di robustezza statistica
#   13. Riepilogo finale + Trading System
# ============================================================

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 1 — Setup & Import                               ║
# ╚══════════════════════════════════════════════════════════╝

import subprocess, sys as _sys
for _pkg in ['missingno', 'optuna']:
    try: __import__(_pkg.replace('-', '_'))
    except ImportError:
        subprocess.check_call([_sys.executable, '-m', 'pip', 'install', _pkg, '-q'])

import sys, math, os, warnings
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno
import torch
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy import stats as _stats

sys.path.append('C:\\python scripts\\ip_2026\\src')
from btcdata_lib import *
from btcfornn_lib import *

warnings.filterwarnings('ignore')
print('✅ Librerie caricate')

"""# Scarico features e creazione Datasets + Feature Engineering"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 2 — Caricamento dati on-chain                    ║
# ╚══════════════════════════════════════════════════════════╝

df = pd.read_excel('ip26_btconchain.xlsx')
df.set_index('date', inplace=True)
df.drop(['MKPRU'], inplace=True, axis=1)   # rimuovi duplicato

print(f'Shape: {df.shape} | {df.index.min().date()} → {df.index.max().date()}')
missingno.matrix(df, figsize=(12, 4), fontsize=10)
plt.title('On-Chain — valori mancanti'); plt.tight_layout(); plt.show()


# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 3 — Normalizzazione on-chain                     ║
# ╚══════════════════════════════════════════════════════════╝
# Z-score rolling 180gg (look-ahead free) + log-transform
# sulle colonne a distribuzione molto asimmetrica.

onchain_log = ['NTRAN','NTRAT','NTRBL','NTREP','TRFEE','TRFUS',
               'TRVOU','TOUTV','ETRAV','ETRVU','DIFF','HRATE']

df_scaled = normalize_onchain_features(
    df,
    [c for c in df.columns if c != 'BTC_CLOSE'],
    window=180,
    log_transform_cols=onchain_log
)
print(f'✅ Normalizzazione completata: {df_scaled.shape[1]} colonne')

"""# Pannello di controllo flags"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 4 — Configurazione                               ║
# ╚══════════════════════════════════════════════════════════╝

# ── Flags di controllo ───────────────────────────────────────
flag_EDA         = 'N'
flag_FREM        = 'Y'   # rimozione feature ridondanti
flag_FIMP        = 'Y'   # feature importance (permutation)
flag_FS          = 'N'   # feature selection dopo il 1° training
flag_SCUT        = 'Y'   # taglia campione da cdate
flag_TRAD        = 'Y'   # trading system
flag_TARGET      = 'L'   # L=log-return / P=prezzo
flag_SMOOTH      = 'Y'   # target smoothing 3gg forward
flag_HP_TUNING   = 'Y'   # Optuna HPT — 'Y' lancia Optuna, 'N' usa best_hp pre-caricati
flag_WF          = 'Y'   # walk-forward validation
flag_SEAS_M      = 'Y'   # dummies mese
flag_SEAS_W      = 'Y'   # dummies settimana
flag_SEAS_D      = 'N'   # dummies giorno settimana
flag_HALV        = 'Y'   # giorni dall'halving

# ── Flags categoria (già definiti in 2b, riconfermati qui) ───
flag_ONCHAIN     = 'Y'
flag_MARKET      = 'N'   # disabilitato: solo on-chain
flag_MACRO       = 'N'   # disabilitato: solo on-chain
flag_SENTIMENT   = 'N'   # disabilitato: solo on-chain
flag_OI          = 'N'   # disabilitato: solo on-chain
flag_TECH        = 'N'   # disabilitato: solo on-chain
flag_REGIME      = 'N'   # disabilitato: solo on-chain
flag_SEASONAL    = 'N'   # disabilitato: solo on-chain

# ── Date split ───────────────────────────────────────────────
# Training : 2017-01-01 → 2025-06-01
# Gap OOS  : 2025-06-01 → 2025-09-01  (3 mesi anti-leakage)
# Test OOS : 2025-09-01 → 2026-01-30  (~5 mesi, ~109 campioni)
cdate            = '2017-01-01'
training_end     = '2025-06-01'
test_start_date  = '2025-09-01'
test_end_date    = '2026-01-30'

# ── Parametri rete ───────────────────────────────────────────
RND_SEED         = 42
n_lags           = 21    # finestra temporale (ottimizzata da Optuna)
n_steps          = 1     # single-step: multi-step è nel target smoothato
val_size         = 0.15  # ultimi 15% del training → validation
scaler_type      = 'MM'

# ── Hyperparametri default (sovrascritti da Optuna se flag_HP_TUNING='Y') ─
m_units          = 64
dropout          = 0.3
learn_rate       = 1e-4
weight_decay     = 1e-4
grad_clip        = 1.0
m_epochs         = 200    # margine extra — early stopping ferma prima se necessario
m_batch_size     = 64
m_patience       = 20     # più pazienza: evita stop prematuro su plateau lunghi
smooth_window    = 3

# ── Walk-forward ─────────────────────────────────────────────
WF_N_FOLDS       = 5
WF_GAP_DAYS      = 7    # gap anti-leakage tra ogni fold train e val

# ── Filtro qualità ensemble ───────────────────────────────────
ENS_MIN_R2       = 0.0
ENS_MIN_DA       = 50.0
ENS_MAX_THEIL    = 1.0

# ── Trading system ───────────────────────────────────────────
init_inv         = 100_000
trans_cost       = 0.001
trad_mode        = 'additive'
take_pr          = 0.0005
stop_l           = -0.0005

# ── Feature da rimuovere (ridondanti) ────────────────────────
feat_rem         = ['MKTCP','BLCHS','MWNUS','NTRAT','TOTBC']

# ── Optuna HPT ───────────────────────────────────────────────
# Trial per modello — differenziati per complessità architetturale.
# I modelli attention hanno search space più complesso (più parametri)
# e richiedono più trial per trovare il minimo.
N_TRIALS         = 40    # modelli base (LSTM, GRU, BI-LSTM)
N_TRIALS_ATT     = 60    # modelli attention (+ATT, SELF-ATT)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
print(f'[ON-CHAIN ONLY] Split: Train 2017-01-01→{training_end} | Gap 3m | Test {test_start_date}→{test_end_date}')

"""# Preprocessing e temporal split"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 5 — Preprocessing & Split temporale              ║
# ╚══════════════════════════════════════════════════════════╝

# ── Carica e prepara dati ────────────────────────────────────
df_btc_op = pd.read_excel('ip26_btconchain.xlsx')
df_btc_op.set_index('date', inplace=True)
df_btc_op.drop(['BTC_VOLUME','BTC_HIGH','BTC_LOW','BTC_CLOSE'], inplace=True, axis=1)

# Rimuovi feature ridondanti
if flag_ONCHAIN == 'Y' and flag_FREM == 'Y':
    to_drop = [c for c in feat_rem if c in df_btc_op.columns]
    df_btc_op.drop(to_drop, inplace=True, axis=1)
    print(f'Rimosse feature ridondanti: {to_drop}')

# ── Normalizzazione on-chain (rolling Z-score 180d + log-transform) ──────────
# Applicata PRIMA del merge delle feature esterne e del sample cut,
# in modo che la finestra rolling 180d abbia abbastanza storia anche
# per le prime osservazioni del campione 2017.
# MKPRU è escluso dalla normalizzazione (è il target grezzo).
if flag_ONCHAIN == 'Y':
    _onchain_cols_to_norm = [
        c for c in df_btc_op.columns
        if c not in ['MKPRU'] and c in [
            'NTRAN','NTRAT','NTRBL','NTREP','TRFEE','TRFUS',
            'TRVOU','TOUTV','ETRAV','ETRVU','CPTRA','CPTRV',
            'DIFF','HRATE','MIREV','AVBLS','NADDU',
        ]
    ]
    _onchain_log_norm = [
        c for c in onchain_log if c in _onchain_cols_to_norm
    ]
    df_btc_op = normalize_onchain_features(
        df_btc_op,
        _onchain_cols_to_norm,
        window=180,
        log_transform_cols=_onchain_log_norm
    )
    print(f'✅ On-chain normalization applicata: {len(_onchain_cols_to_norm)} colonne')

# Nessun merge di feature esterne — modello on-chain only

# Taglio campione
if flag_SCUT == 'Y':
    df_btc_op = df_btc_op.loc[cdate:]

# Log-return target
df_btc_op['MKPRU_logreturns'] = np.log(df_btc_op['MKPRU'] / df_btc_op['MKPRU'].shift(1)).fillna(0)
target_type = 'logreturns'

# ── Feature tecniche su MKPRU ────────────────────────────────
# NOTA LOOK-AHEAD: tutte le feature tecniche includono price_t o lr_t
# del giorno corrente se calcolate al "raw". Applichiamo shift(1) a tutto
# in modo che al giorno t il modello veda la feature calcolata al giorno t-1.
# Eccezione: LR_LAG1/2/3 sono già shift(1)/shift(2)/shift(3) per costruzione.
if flag_TECH == 'Y':
    price   = df_btc_op['MKPRU']
    lr      = np.log(price/price.shift(1))   # log-return grezzo

    # RSI14: calcolato su delta al giorno t, poi laggato
    delta   = price.diff()
    gain    = delta.clip(lower=0).rolling(14).mean()
    loss    = (-delta.clip(upper=0)).rolling(14).mean()
    df_btc_op['RSI14']      = (100 - (100/(1 + gain/(loss+1e-10)))).shift(1)

    # MACD: EWM su prezzi passati, ma include il giorno corrente → shift(1)
    ema12   = price.ewm(span=12, adjust=False).mean()
    ema26   = price.ewm(span=26, adjust=False).mean()
    macd    = ema12 - ema26
    macd_signal = macd.ewm(span=9, adjust=False).mean()
    df_btc_op['MACD']       = macd.shift(1)
    df_btc_op['MACD_SIGNAL']= macd_signal.shift(1)
    df_btc_op['MACD_HIST']  = (macd - macd_signal).shift(1)

    # Bollinger Bands: rolling(20) include giorno corrente → shift(1)
    bb_mid  = price.rolling(20).mean()
    bb_std  = price.rolling(20).std()
    df_btc_op['BB_PCT']     = ((price-(bb_mid-2*bb_std))/(4*bb_std+1e-10)).shift(1)
    df_btc_op['BB_WIDTH']   = ((4*bb_std)/(bb_mid+1e-10)).shift(1)

    # ATR14: include |ΔP_t| → shift(1)
    df_btc_op['ATR14']      = price.diff().abs().rolling(14).mean().shift(1)

    # Stochastic: usa price_t nel numeratore → shift(1)
    roll_min= price.rolling(14).min()
    roll_max= price.rolling(14).max()
    stoch_k = 100*(price-roll_min)/(roll_max-roll_min+1e-10)
    df_btc_op['STOCH_K']    = stoch_k.shift(1)
    df_btc_op['STOCH_D']    = stoch_k.rolling(3).mean().shift(1)

    # Momentum rolling: include lr_t → shift(1)
    for w in [5,10,20]:
        df_btc_op[f'LR_{w}D']  = lr.rolling(w).sum().shift(1)
        df_btc_op[f'VOL_{w}D'] = lr.rolling(w).std().shift(1)

    # Lagged log-returns: già look-ahead free per costruzione
    for lag in [1,2,3]:
        df_btc_op[f'LR_LAG{lag}'] = lr.shift(lag)

    # Distanza da MA: (price_t - MA_t)/MA_t include price_t → shift(1)
    for w in [7,21,50]:
        ma = price.rolling(w).mean()
        df_btc_op[f'DIST_MA{w}'] = ((price-ma)/(ma+1e-10)).shift(1)

# ── Feature di regime ────────────────────────────────────────
# Tutte le regime feature usano price_t o lr_t del giorno corrente.
# Applichiamo shift(1) per garantire look-ahead free al giorno t.
if flag_REGIME == 'Y':
    price   = df_btc_op['MKPRU']
    lr_r    = np.log(price/price.shift(1))

    # ATH_DIST: expanding max include price_t → shift(1)
    ath     = price.expanding().max()
    _ath_dist = (price-ath)/(ath+1e-10)
    df_btc_op['ATH_DIST']     = _ath_dist.shift(1)
    df_btc_op['ATH_DIST_MA']  = _ath_dist.rolling(30).mean().shift(1)

    # VOL_REGIME: rolling std su lr che include lr_t → shift(1)
    df_btc_op['VOL_REGIME']   = (lr_r.rolling(30).std()/(lr_r.rolling(90).std()+1e-10)).shift(1)

    # TREND_60D: log(price_t/price_{t-60}) include price_t → shift(1)
    df_btc_op['TREND_60D']    = np.log(price/price.shift(60)).shift(1)

    # ABOVE_MA200: price_t > MA200_t → shift(1)
    df_btc_op['ABOVE_MA200']  = (price>price.rolling(200).mean()).astype(float).shift(1)

# ── Stagionali e halving ─────────────────────────────────────
if flag_SEASONAL == 'Y':
    df_btc_op = add_seasonal_features(df_btc_op, flag_SEAS_M, flag_SEAS_W, flag_SEAS_D)
    if flag_HALV == 'Y':
        df_btc_op = add_halving_features(df_btc_op)

# ── Pulizia numerica ─────────────────────────────────────────
df_btc_op.replace([np.inf, -np.inf], np.nan, inplace=True)

# Diagnostica: quali colonne hanno NaN e quanti?
_nan_counts = df_btc_op.isna().sum()
_nan_cols = _nan_counts[_nan_counts > 0].sort_values(ascending=False)
if len(_nan_cols) > 0:
    print(f'\n📊 NaN per colonna prima della pulizia (top 15):')
    print(_nan_cols.head(15).to_string())

# Step 1: ffill/bfill standard per gap weekend/festivi
df_btc_op.ffill(limit=5, inplace=True)
df_btc_op.bfill(limit=5, inplace=True)

# Step 2: nessuna feature esterna — skip imputation

# Step 3: per le feature tecniche e di regime (rolling su MKPRU),
# il warm-up produce NaN solo nelle primissime righe — bfill limitato le copre.
df_btc_op.bfill(limit=200, inplace=True)

# Step 4: dropna finale — dovrebbe eliminare pochissime righe (solo warm-up iniziale)
_n_before = len(df_btc_op)
df_btc_op.dropna(inplace=True)
_n_after = len(df_btc_op)
print(f'\n✅ dropna: eliminate {_n_before - _n_after} righe | rimaste {_n_after}')
print(f'   Campione effettivo: {df_btc_op.index.min().date()} → {df_btc_op.index.max().date()}')

# Outlier clipping: µ e σ calcolati SOLO sul training set (leak-free).
# Il test set viene clippato agli stessi bounds del training.
_clip_dates = df_btc_op.index
_train_mask_clip = (_clip_dates >= pd.Timestamp(cdate)) & \
                   (_clip_dates < pd.Timestamp(training_end))
for col in df_btc_op.select_dtypes(include=[np.number]).columns:
    if col == 'MKPRU': continue
    m = df_btc_op.loc[_train_mask_clip, col].mean()
    s = df_btc_op.loc[_train_mask_clip, col].std()
    if s > 0:
        df_btc_op[col] = df_btc_op[col].clip(m - 5*s, m + 5*s)

# Rimuovi MKPRU_logreturns dall'input (leakage diretto sul target)
data_model = df_btc_op.drop(columns=['MKPRU_logreturns'])

print(f'Feature totali: {data_model.shape[1]-1} + target MKPRU')

# ── Costruzione finestre LSTM ─────────────────────────────────
X, y    = prepare_lstm_input(data_model, target='MKPRU', lags=n_lags,
                              target_type=target_type, flagDIR='N')
seq_dates = df_btc_op.index[n_lags:]
total_N   = len(X)

# ── Split temporale ───────────────────────────────────────────
# Training : fino a training_end
# Gap      : training_end → test_start_date  (3 mesi, escluso)
# Test OOS : test_start_date → test_end_date
tv_mask = (seq_dates >= pd.Timestamp(cdate)) & (seq_dates < pd.Timestamp(training_end))
ts_mask = (seq_dates >= pd.Timestamp(test_start_date)) & (seq_dates <= pd.Timestamp(test_end_date))

X_trainval = X[tv_mask];  y_trainval = y[tv_mask]
X_test     = X[ts_mask];  y_test     = y[ts_mask]
test_dates_idx     = seq_dates[ts_mask]
trainval_dates_idx = seq_dates[tv_mask]

# Validation: ultimi val_size% del training
val_n  = int(val_size * len(X_trainval))
train_n= len(X_trainval) - val_n
X_train = X_trainval[:train_n];  y_train = y_trainval[:train_n]
X_val   = X_trainval[train_n:];  y_val   = y_trainval[train_n:]
train_dates_idx = trainval_dates_idx[:train_n]
val_dates_idx   = trainval_dates_idx[train_n:]

# ── Scaling (fit SOLO su X_train) ────────────────────────────
scaler_X = MinMaxScaler() if scaler_type=='MM' else StandardScaler()
scaler_y = MinMaxScaler() if scaler_type=='MM' else StandardScaler()

X_train_sc = scaler_X.fit_transform(X_train.reshape(-1,X_train.shape[2])).reshape(X_train.shape)
X_val_sc   = scaler_X.transform(X_val.reshape(-1,X_val.shape[2])).reshape(X_val.shape)
X_test_sc  = scaler_X.transform(X_test.reshape(-1,X_test.shape[2])).reshape(X_test.shape)
y_train_sc = scaler_y.fit_transform(y_train.reshape(-1,1))
y_val_sc   = scaler_y.transform(y_val.reshape(-1,1))
y_test_sc  = scaler_y.transform(y_test.reshape(-1,1))

input_shape = (n_lags, X_train_sc.shape[2])

# Prezzi t-1 per ricostruzione da log-return.
# price_t1_all[i] = MKPRU[i + n_lags - 1] = prezzo del giorno precedente
# alla previsione i, allineato a seq_dates = df_btc_op.index[n_lags:].
# Usiamo seq_dates per indicizzare cosi il mapping e sempre corretto
# indipendentemente da n_lags.
def _build_price_t1(df_btc_op, seq_dates, tv_mask, ts_mask, train_n):
    """Restituisce (train_price_t1, val_price_t1, test_price_t1)."""
    ap     = df_btc_op['MKPRU'].values
    # seq_dates[i] corrisponde a df_btc_op.index[n_lags + i]
    # Il prezzo t-1 di seq_dates[i] e df_btc_op.index[n_lags + i - 1]
    # ovvero ap[n_lags + i - 1]. Poiche seq_dates = df_btc_op.index[n_lags:],
    # possiamo trovare l'indice intero tramite searchsorted.
    df_idx = df_btc_op.index
    pos    = np.array([df_idx.get_loc(d) - 1 for d in seq_dates])
    pt1_all      = ap[pos]
    train_pt1    = pt1_all[tv_mask][:train_n]
    val_pt1      = pt1_all[tv_mask][train_n:]
    test_pt1     = pt1_all[ts_mask]
    return train_pt1, val_pt1, test_pt1

train_price_t1, val_price_t1, test_price_t1 = _build_price_t1(
    df_btc_op, seq_dates, tv_mask, ts_mask, train_n)

set_seeds(RND_SEED)

print(f'\n✅ Split completato:')
print(f'   Train:      {train_n:5d} campioni ({train_dates_idx.min().date()} → {train_dates_idx.max().date()})')
print(f'   Validation: {val_n:5d} campioni ({val_dates_idx.min().date()} → {val_dates_idx.max().date()})')
print(f'   Gap OOS:    {training_end} → {test_start_date} (escluso)')
print(f'   Test OOS:   {len(X_test):5d} campioni ({test_dates_idx.min().date()} → {test_dates_idx.max().date()})')
print(f'   Input shape: {input_shape}')

"""# Walk-forward cross validation con gap"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 5b — Walk-Forward Validation con gap             ║
# ╚══════════════════════════════════════════════════════════╝
# Schema expanding window con gap di WF_GAP_DAYS tra train e val.
# Ogni fold: Train[0:T_i] → skip WF_GAP_DAYS → Val[T_i+gap : T_i+gap+step]
# Questo previene che informazione recente nel training "contamini"
# la validation attraverso feature rolling e momentum.

def build_wf_folds(X_tv, y_tv, n_folds, gap_days, val_size_ratio,
                   scaler_type='MM'):
    """
    Costruisce fold di walk-forward con gap anti-leakage.

    Parametri
    ---------
    gap_days : int — giorni da saltare tra fine training e inizio validation
    """
    tv_n    = len(X_tv)
    base_tr = int((1-val_size_ratio) * tv_n)
    step    = (tv_n - base_tr - gap_days) // n_folds

    folds = []
    for i in range(n_folds):
        tr_end  = base_tr + i * step
        val_st  = tr_end + gap_days          # salta il gap
        val_end = val_st + step
        if val_end > tv_n: break

        X_tr_f = X_tv[:tr_end]
        y_tr_f = y_tv[:tr_end]
        X_vl_f = X_tv[val_st:val_end]
        y_vl_f = y_tv[val_st:val_end]

        sc_X = MinMaxScaler() if scaler_type=='MM' else StandardScaler()
        sc_y = MinMaxScaler() if scaler_type=='MM' else StandardScaler()
        X_tr_sc = sc_X.fit_transform(X_tr_f.reshape(-1,X_tr_f.shape[2])).reshape(X_tr_f.shape)
        X_vl_sc = sc_X.transform(X_vl_f.reshape(-1,X_vl_f.shape[2])).reshape(X_vl_f.shape)
        y_tr_sc = sc_y.fit_transform(y_tr_f.reshape(-1,1))
        y_vl_sc = sc_y.transform(y_vl_f.reshape(-1,1))

        folds.append({
            'fold': i+1,
            'tr_end': tr_end, 'val_st': val_st, 'val_end': val_end,
            'X_tr': X_tr_sc, 'y_tr': y_tr_sc,
            'X_val': X_vl_sc,'y_val': y_vl_sc,
            'sc_y': sc_y,
        })
        print(f'  Fold {i+1}: train[0:{tr_end}] → gap {gap_days}d → val[{val_st}:{val_end}]')

    return folds

# NOTA: wf_folds qui usa n_lags del consensus (ricostruito in cella 6 dopo Optuna).
# In cella 5b costruiamo un WF preliminare con n_lags corrente che verrà
# sovrascritto dalla cella 6 post-consensus. Ogni modello usa poi _wf_XXX
# costruito da build_model_split() con il proprio n_lags ottimale.
if flag_WF == 'Y':
    print(f'\n🔄 Walk-Forward preliminare ({WF_N_FOLDS} fold, gap={WF_GAP_DAYS}d):')
    wf_folds = build_wf_folds(X_trainval, y_trainval,
                               WF_N_FOLDS, WF_GAP_DAYS, val_size, scaler_type)
    print(f'✅ {len(wf_folds)} fold pronti (verrà ricalcolato post-consensus)')
else:
    wf_folds = []

"""# Hyperparameter Tuning con Optuna"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 6 — Hyperparameter Tuning (Optuna)               ║
# ╚══════════════════════════════════════════════════════════╝
# TPE separato per ogni architettura.
# Obiettivo: min MSE_val (leak-free).
# Se n_lags ottimale != consensus, ricostruisce il dataset.

# ── Parametri ottimali pre-caricati (da ultima run Optuna) ───
# Se flag_HP_TUNING='Y' questi vengono sovrascritti da Optuna.
# Se flag_HP_TUNING='N' vengono usati direttamente — skip Optuna.
best_hp = {
    'LSTM':        {'m_units': 48,  'dropout': 0.344, 'learn_rate': 7.25e-4,
                    'weight_decay': 3.03e-4, 'n_lags': 44, 'best_mse_val': 0.002115,
                    'dir_alpha': 0.3},
    'GRU':         {'m_units': 64,  'dropout': 0.320, 'learn_rate': 7.70e-4,
                    'weight_decay': 1.09e-4, 'n_lags': 23, 'best_mse_val': 0.002131,
                    'dir_alpha': 0.3},
    'BI-LSTM':     {'m_units': 32,  'dropout': 0.474, 'learn_rate': 7.77e-4,
                    'weight_decay': 8.75e-5, 'n_lags': 42, 'best_mse_val': 0.002060,
                    'dir_alpha': 0.3},
    'LSTM-ATT':    {'m_units': 48,  'dropout': 0.347, 'learn_rate': 1.62e-4,
                    'weight_decay': 2.31e-4, 'n_lags': 42, 'best_mse_val': 0.001948,
                    'dir_alpha': 0.3},
    'GRU-ATT':     {'m_units': 48,  'dropout': 0.529, 'learn_rate': 9.32e-4,
                    'weight_decay': 1.47e-3, 'n_lags': 42, 'best_mse_val': 0.001998,
                    'dir_alpha': 0.3},
    'BI-LSTM-ATT': {'m_units': 48,  'dropout': 0.445, 'learn_rate': 9.51e-4,
                    'weight_decay': 1.65e-4, 'n_lags': 42, 'best_mse_val': 0.001966,
                    'dir_alpha': 0.3},
    'SELF-ATT':    {'m_units': 32,  'dropout': 0.308, 'learn_rate': 3.82e-4,
                    'weight_decay': 1.82e-4, 'n_lags': 42, 'best_mse_val': 0.001948,
                    'dir_alpha': 0.3},
}

import torch.nn as nn
import torch.nn.functional as F

class DirectionalLoss(nn.Module):
    """
    Loss composita: (1-alpha)*MSE + alpha*penalità_direzione.
    Penalizza quando pred e true hanno segno opposto,
    proporzionalmente alla magnitudine dell'errore.
    alpha=0 → MSE pura | alpha=1 → solo direzione
    """
    def __init__(self, alpha=0.3):
        super().__init__()
        self.alpha = alpha

    def forward(self, pred, true):
        mse_loss = F.mse_loss(pred, true)
        dir_loss = torch.mean(torch.clamp(-pred * true, min=0))
        return (1.0 - self.alpha) * mse_loss + self.alpha * dir_loss


def train_regularized(model, X_tr, y_tr, X_val, y_val,
                       epochs, batch_size, lr, patience,
                       device, wd=0.0, clip=None,
                       warmup_epochs=5, accum_steps=2,
                       es_smooth_w=3, dir_alpha=0.0):
    """
    Training loop con:
    - Linear warmup del lr nei primi warmup_epochs epoch
      (parte da lr/10, sale linearmente a lr) → elimina i picchi iniziali
    - Gradient accumulation ogni accum_steps batch
      (simula batch effettivo = batch_size * accum_steps) → meno varianza
    - ReduceLROnPlateau su val loss EMA → lr scheduling stabile
    - Early stopping su val loss EMA (media mobile es_smooth_w epoch)
      invece del valore raw → meno sensibile ai picchi isolati
    - Training loss e val loss entrambe in eval() → grafico comparabile
    - dir_alpha > 0: usa DirectionalLoss invece di MSE pura
    """
    import torch.nn as nn
    model  = model.to(device)
    opt    = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=wd)

    # Warmup: LinearLR da lr/10 a lr in warmup_epochs epoch
    warmup_sched = torch.optim.lr_scheduler.LinearLR(
        opt, start_factor=0.1, end_factor=1.0,
        total_iters=warmup_epochs)
    # Post-warmup: ReduceLROnPlateau
    plateau_sched = torch.optim.lr_scheduler.ReduceLROnPlateau(
        opt, mode='min', factor=0.5, patience=10, min_lr=1e-6)

    # Loss: DirectionalLoss se dir_alpha > 0, altrimenti MSE pura
    crit = DirectionalLoss(alpha=dir_alpha) if dir_alpha > 0 else nn.MSELoss()
    # Val loss sempre in MSE pura per confronto equo tra modelli
    mse_crit = nn.MSELoss()

    X_tr_t = torch.tensor(X_tr,  dtype=torch.float32).to(device)
    y_tr_t = torch.tensor(y_tr,  dtype=torch.float32).to(device)
    X_vl_t = torch.tensor(X_val, dtype=torch.float32).to(device)
    y_vl_t = torch.tensor(y_val, dtype=torch.float32).to(device)

    best_val, best_state, wait = float('inf'), None, 0
    tr_losses, vl_losses = [], []
    vl_history = []   # buffer per EMA early stopping

    for epoch in range(1, epochs + 1):

        # ── Forward + backward con gradient accumulation ──────
        model.train()
        perm = torch.randperm(len(X_tr_t), device=device)
        opt.zero_grad()
        step_count = 0
        for s in range(0, len(X_tr_t), batch_size):
            idx  = perm[s:s + batch_size]
            loss = crit(model(X_tr_t[idx]), y_tr_t[idx])
            # Normalizza per accum_steps così la magnitudine del gradiente
            # è indipendente dal numero di passi di accumulo
            (loss / accum_steps).backward()
            step_count += 1
            if step_count % accum_steps == 0 or \
               s + batch_size >= len(X_tr_t):
                if clip:
                    nn.utils.clip_grad_norm_(model.parameters(), clip)
                opt.step()
                opt.zero_grad()

        # ── Valutazione in eval() ─────────────────────────────
        model.eval()
        with torch.no_grad():
            # Val loss sempre MSE per early stopping coerente
            vl      = mse_crit(model(X_vl_t), y_vl_t).item()
            tr_eval = mse_crit(model(X_tr_t), y_tr_t).item()

        tr_losses.append(tr_eval)
        vl_losses.append(vl)

        # ── Scheduler step ────────────────────────────────────
        if epoch <= warmup_epochs:
            warmup_sched.step()
        else:
            plateau_sched.step(vl)

        # ── Val loss smoothata per early stopping ─────────────
        # Usiamo la media delle ultime es_smooth_w val loss
        # per non reagire ai picchi isolati
        vl_history.append(vl)
        vl_smooth = float(sum(vl_history[-es_smooth_w:]) /
                          len(vl_history[-es_smooth_w:]))

        if epoch % 20 == 0:
            current_lr = opt.param_groups[0]['lr']
            print(f'  ep {epoch:3d} | tr={tr_eval:.6f} | '
                  f'vl={vl:.6f} | vl_ema={vl_smooth:.6f} | '
                  f'lr={current_lr:.2e}')

        # ── Early stopping su vl_smooth ───────────────────────
        if vl_smooth < best_val - 1e-6:
            best_val   = vl_smooth
            best_state = {k: v.clone() for k, v in model.state_dict().items()}
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f'  Early stop @ epoch {epoch} '
                      f'(vl_ema={vl_smooth:.6f})')
                break


    if best_state:
        model.load_state_dict(best_state)
    return model, tr_losses, vl_losses

def _train(model, Xtr, ytr, Xval, yval):
    """Dispatcher: usa sempre train_regularized con i parametri correnti."""
    hp        = best_hp.get(getattr(model,'_name',''), {})
    lr        = hp.get('learn_rate',  learn_rate)
    wd        = hp.get('weight_decay', weight_decay)
    dir_alpha = hp.get('dir_alpha',    0.0)
    return train_regularized(model, Xtr, ytr, Xval, yval,
                              m_epochs, m_batch_size, lr, m_patience,
                              device, wd=wd, clip=grad_clip,
                              dir_alpha=dir_alpha)

if flag_HP_TUNING == 'Y':
    import optuna
    from optuna.pruners import MedianPruner
    from collections import Counter as _Counter
    optuna.logging.set_verbosity(optuna.logging.WARNING)

    _build_fns = {
        'LSTM':        build_lstm_model,
        'GRU':         build_gru_model,
        'BI-LSTM':     build_bilstm_model,
        'LSTM-ATT':    build_lstm_attention_model,
        'GRU-ATT':     build_gru_attention_model,
        'BI-LSTM-ATT': build_bilstm_attention_model,
        'SELF-ATT':    build_self_attention_model,
    }
    _att_models  = {'LSTM-ATT', 'GRU-ATT', 'BI-LSTM-ATT', 'SELF-ATT'}
    _OPT_EPOCHS  = 150
    _OPT_PAT     = 20
    _WF_OPT_FOLDS= 3
    _WF_OPT_GAP  = 7
    _fold_cache  = {}

    def _build_optuna_folds(X_tv, y_tv, n_folds, gap, val_ratio, sc_type='MM'):
        tv_n    = len(X_tv)
        opt_n   = int(tv_n * 0.80)
        base_tr = int((1 - val_ratio) * opt_n)
        step    = max(1, (opt_n - base_tr - gap) // n_folds)
        folds   = []
        for i in range(n_folds):
            tr_end  = base_tr + i * step
            val_st  = tr_end + gap
            val_end = val_st + step
            if val_end > opt_n:
                break
            Xtrf = X_tv[:tr_end];       ytrf = y_tv[:tr_end]
            Xvlf = X_tv[val_st:val_end]; yvlf = y_tv[val_st:val_end]
            scX  = MinMaxScaler() if sc_type=='MM' else StandardScaler()
            scY  = MinMaxScaler() if sc_type=='MM' else StandardScaler()
            Xtr_sc = scX.fit_transform(Xtrf.reshape(-1,Xtrf.shape[2])).reshape(Xtrf.shape)
            Xvl_sc = scX.transform(Xvlf.reshape(-1,Xvlf.shape[2])).reshape(Xvlf.shape)
            ytr_sc = scY.fit_transform(ytrf.reshape(-1,1))
            yvl_sc = scY.transform(yvlf.reshape(-1,1))
            folds.append({'X_tr':Xtr_sc,'y_tr':ytr_sc,'X_val':Xvl_sc,'y_val':yvl_sc})
        return folds

    def _get_optuna_folds(nl):
        if nl in _fold_cache:
            return _fold_cache[nl]
        if nl == n_lags:
            _Xtv, _ytv = X_trainval, y_trainval
        else:
            _Xa, _ya = prepare_lstm_input(data_model, target='MKPRU',
                                           lags=nl, target_type=target_type, flagDIR='N')
            _sd  = df_btc_op.index[nl:]
            _tvm = (_sd>=pd.Timestamp(cdate))&(_sd<pd.Timestamp(training_end))
            _Xtv = _Xa[_tvm]; _ytv = _ya[_tvm]
        folds = _build_optuna_folds(_Xtv, _ytv, _WF_OPT_FOLDS,
                                     _WF_OPT_GAP, val_size, scaler_type)
        _fold_cache[nl] = (_Xtv, _ytv, folds)
        return _Xtv, _ytv, folds

    def _make_objective(build_fn, mname):
        is_att = mname in _att_models
        def _obj(trial):
            hp = {
                'm_units':      trial.suggest_int('m_units',
                                    32, 64 if is_att else 128, step=16),
                'dropout':      trial.suggest_float('dropout',
                                    0.2 if is_att else 0.1,
                                    0.6 if is_att else 0.5),
                'learn_rate':   trial.suggest_float('learn_rate', 1e-5, 5e-4, log=True),
                'weight_decay': trial.suggest_float('weight_decay',
                                    1e-5 if is_att else 1e-6,
                                    1e-2 if is_att else 1e-3, log=True),
                'n_lags':       trial.suggest_int('n_lags', 14, 90),
                'dir_alpha':    trial.suggest_float('dir_alpha', 0.1, 0.5),
            }
            _Xtv, _ytv, folds = _get_optuna_folds(hp['n_lags'])
            if not folds:
                return float('inf')
            fold_mses = []
            for fi, fold in enumerate(folds):
                set_seeds(RND_SEED + fi)
                sh_f = (hp['n_lags'], fold['X_tr'].shape[2])
                m    = build_fn(sh_f, net_units=hp['m_units'], act_f='tanh',
                                dropout=hp['dropout'], for_steps=1,
                                opt_rate=hp['learn_rate'], opt_clip=None, metric='mse')
                _, _, vl_losses = train_regularized(
                    m, fold['X_tr'], fold['y_tr'],
                    fold['X_val'], fold['y_val'],
                    epochs=_OPT_EPOCHS, batch_size=m_batch_size,
                    lr=hp['learn_rate'], patience=_OPT_PAT,
                    device=device, wd=hp['weight_decay'],
                    clip=grad_clip, dir_alpha=hp['dir_alpha'])
                fold_mses.append(min(vl_losses) if vl_losses else float('inf'))
                trial.report(float(np.mean(fold_mses)), step=fi)
                if trial.should_prune():
                    raise optuna.exceptions.TrialPruned()
            return float(np.mean(fold_mses))
        return _obj

    print('='*60)
    print(f'  OPTUNA HPT — WF {_WF_OPT_FOLDS}-fold | n_lags∈[14,90]')
    print(f'  {N_TRIALS} trial (base) / {N_TRIALS_ATT} trial (attention)')
    print('='*60)

    n_models  = len(_build_fns)
    fig, axes = plt.subplots(2, 4, figsize=(24, 8))
    axes_flat = axes.flatten()

    for i, (mname, build_fn) in enumerate(_build_fns.items()):
        n_tr_this = N_TRIALS_ATT if mname in _att_models else N_TRIALS
        print(f'\n[{i+1}/{n_models}] {mname} ({n_tr_this} trial)...')
        pruner = MedianPruner(n_startup_trials=10, n_warmup_steps=1, interval_steps=1)
        study  = optuna.create_study(
            direction='minimize',
            sampler=optuna.samplers.TPESampler(seed=RND_SEED+i),
            pruner=pruner)
        study.optimize(_make_objective(build_fn, mname),
                       n_trials=n_tr_this, show_progress_bar=False)
        bp = study.best_params
        best_hp[mname] = {**bp, 'best_mse_val': study.best_value}
        n_pruned    = sum(1 for t in study.trials
                          if t.state == optuna.trial.TrialState.PRUNED)
        n_completed = sum(1 for t in study.trials
                          if t.state == optuna.trial.TrialState.COMPLETE)
        print(f'  Best WF-MSE={study.best_value:.6f} | '
              f'units={bp["m_units"]} drop={bp["dropout"]:.3f} '
              f'lr={bp["learn_rate"]:.2e} n_lags={bp["n_lags"]} '
              f'dir_α={bp["dir_alpha"]:.2f}')
        print(f'  Completati={n_completed} | Potati={n_pruned}')
        tdf    = study.trials_dataframe()
        tdf_ok = tdf[tdf['state'] == 'COMPLETE']
        ax     = axes_flat[i]
        ax.plot(tdf_ok.index, tdf_ok['value'], 'o-', markersize=4, alpha=0.7)
        ax.axhline(study.best_value, color='green', ls='--',
                   label=f'Best: {study.best_value:.5f}')
        ax.set_title(mname, fontweight='bold')
        ax.set_xlabel('Trial'); ax.legend(fontsize=8); ax.grid(alpha=0.3)

    for j in range(n_models, len(axes_flat)):
        axes_flat[j].set_visible(False)
    plt.suptitle('Optuna — Convergenza HPT (WF objective)',
                 fontsize=13, fontweight='bold')
    plt.tight_layout(); plt.show()

    print('\n  Parametri ottimali:')
    print(f'  {"Modello":<12} {"units":>6} {"dropout":>8} {"lr":>10} '
          f'{"wd":>10} {"n_lags":>7} {"dir_α":>6} {"WF-MSE":>10}')
    for mn, hp in best_hp.items():
        print(f'  {mn:<12} {hp["m_units"]:>6} {hp["dropout"]:>8.3f} '
              f'{hp["learn_rate"]:>10.2e} {hp["weight_decay"]:>10.2e} '
              f'{hp["n_lags"]:>7} {hp.get("dir_alpha",0):>6.2f} '
              f'{hp.get("best_mse_val",float("nan")):>10.6f}')

# ── Consensus n_lags per il test set comune ─────────────────
# Il test set usa il consensus (moda) per avere date allineate tra modelli.
# Ogni modello viene però ADDESTRATO con il proprio n_lags ottimale
# (vedi build_model_split() usata nelle celle 7-13).
from collections import Counter
lags_list = [best_hp[m]['n_lags'] for m in best_hp if 'n_lags' in best_hp[m]]
n_lags_cons = Counter(lags_list).most_common(1)[0][0]
print(f'\n⚠️  n_lags per modello: {dict(zip(best_hp.keys(),lags_list))} → consensus={n_lags_cons}')

# Ricostruisce sempre con consensus per garantire test set allineato
if n_lags_cons != n_lags:
    n_lags = n_lags_cons

X, y   = prepare_lstm_input(data_model, target='MKPRU', lags=n_lags,
                             target_type=target_type, flagDIR='N')
seq_dates = df_btc_op.index[n_lags:]
tv_mask = (seq_dates>=pd.Timestamp(cdate))&(seq_dates<pd.Timestamp(training_end))
ts_mask = (seq_dates>=pd.Timestamp(test_start_date))&(seq_dates<=pd.Timestamp(test_end_date))
X_trainval=X[tv_mask]; y_trainval=y[tv_mask]
X_test=X[ts_mask];     y_test=y[ts_mask]
test_dates_idx=seq_dates[ts_mask]; trainval_dates_idx=seq_dates[tv_mask]
val_n=int(val_size*len(X_trainval)); train_n=len(X_trainval)-val_n
X_train=X_trainval[:train_n]; y_train=y_trainval[:train_n]
X_val=X_trainval[train_n:];   y_val=y_trainval[train_n:]
train_dates_idx=trainval_dates_idx[:train_n]; val_dates_idx=trainval_dates_idx[train_n:]
scaler_X=MinMaxScaler(); scaler_y=MinMaxScaler()
X_train_sc=scaler_X.fit_transform(X_train.reshape(-1,X_train.shape[2])).reshape(X_train.shape)
X_val_sc=scaler_X.transform(X_val.reshape(-1,X_val.shape[2])).reshape(X_val.shape)
X_test_sc=scaler_X.transform(X_test.reshape(-1,X_test.shape[2])).reshape(X_test.shape)
y_train_sc=scaler_y.fit_transform(y_train.reshape(-1,1))
y_val_sc=scaler_y.transform(y_val.reshape(-1,1))
y_test_sc=scaler_y.transform(y_test.reshape(-1,1))
input_shape=(n_lags,X_train_sc.shape[2])
train_price_t1, val_price_t1, test_price_t1 = _build_price_t1(
    df_btc_op, seq_dates, tv_mask, ts_mask, train_n)
if flag_WF=='Y':
    wf_folds=build_wf_folds(X_trainval,y_trainval,WF_N_FOLDS,WF_GAP_DAYS,val_size,scaler_type)
print(f'  Split consensus | shape={input_shape} | test={len(X_test)}')

# ── Helper: costruisce split per un modello con il suo n_lags specifico ──
def build_model_split(mname):
    """
    Restituisce tutti gli array necessari per addestrare e valutare un modello
    con il proprio n_lags ottimale, incluse date e prezzi specifici.
    Se n_lags == consensus, riusa gli array già calcolati.
    """
    ml = best_hp[mname]['n_lags']
    if ml == n_lags:
        return (X_train_sc, X_val_sc, X_test_sc,
                y_train_sc, y_val_sc, y_test_sc,
                scaler_y, input_shape, wf_folds,
                train_price_t1, val_price_t1, test_price_t1,
                lr_train_true, lr_val_true, lr_test_true,
                lr_train_sm, lr_val_sm, lr_test_sm,
                train_dates_idx, test_dates_idx,
                y_train_prices, y_test_prices)
    # n_lags specifico — ricostruisce tensore e split
    _X, _y = prepare_lstm_input(data_model, target='MKPRU', lags=ml,
                                 target_type=target_type, flagDIR='N')
    _sd = df_btc_op.index[ml:]
    _tvm = (_sd>=pd.Timestamp(cdate))&(_sd<pd.Timestamp(training_end))
    _tsm = (_sd>=pd.Timestamp(test_start_date))&(_sd<=pd.Timestamp(test_end_date))
    _Xtv=_X[_tvm]; _ytv=_y[_tvm]
    _Xts=_X[_tsm]; _yts=_y[_tsm]
    _vn=int(val_size*len(_Xtv)); _tn=len(_Xtv)-_vn
    _Xtr=_Xtv[:_tn]; _ytr=_ytv[:_tn]
    _Xvl=_Xtv[_tn:]; _yvl=_ytv[_tn:]
    _sX=MinMaxScaler(); _sy=MinMaxScaler()
    _Xtr_sc=_sX.fit_transform(_Xtr.reshape(-1,_Xtr.shape[2])).reshape(_Xtr.shape)
    _Xvl_sc=_sX.transform(_Xvl.reshape(-1,_Xvl.shape[2])).reshape(_Xvl.shape)
    _Xts_sc=_sX.transform(_Xts.reshape(-1,_Xts.shape[2])).reshape(_Xts.shape)
    _ytr_sc=_sy.fit_transform(_ytr.reshape(-1,1))
    _yvl_sc=_sy.transform(_yvl.reshape(-1,1))
    _yts_sc=_sy.transform(_yts.reshape(-1,1))
    _sh=(ml,_Xtr_sc.shape[2])
    _wf=build_wf_folds(_Xtv,_ytv,WF_N_FOLDS,WF_GAP_DAYS,val_size,scaler_type) if flag_WF=='Y' else []
    _tsd_all=_sd[_tvm]; _trd=_tsd_all[:_tn]; _tsd=_sd[_tsm]
    _tr_pt1,_vl_pt1,_ts_pt1=_build_price_t1(df_btc_op,_sd,_tvm,_tsm,_tn)
    # lr_true e prezzi specifici per questo n_lags
    _lr_tr = _sy.inverse_transform(_ytr_sc).flatten()
    _lr_vl = _sy.inverse_transform(_yvl_sc).flatten()
    _lr_ts = _sy.inverse_transform(_yts_sc).flatten()
    _lr_tr_sm = _smooth_lr(_lr_tr, smooth_window) if flag_SMOOTH=='Y' else None
    _lr_vl_sm = _smooth_lr(_lr_vl, smooth_window) if flag_SMOOTH=='Y' else None
    _lr_ts_sm = _smooth_lr(_lr_ts, smooth_window) if flag_SMOOTH=='Y' else None
    _ytr_prices = df_btc_op.loc[_trd, 'MKPRU'].values
    _yts_prices = df_btc_op.loc[_tsd, 'MKPRU'].values
    print(f'  [{mname}] n_lags={ml} → shape={_sh} | train={_tn} | test={len(_Xts)}')
    return (_Xtr_sc,_Xvl_sc,_Xts_sc,_ytr_sc,_yvl_sc,_yts_sc,_sy,_sh,_wf,
            _tr_pt1,_vl_pt1,_ts_pt1,
            _lr_tr,_lr_vl,_lr_ts,_lr_tr_sm,_lr_vl_sm,_lr_ts_sm,
            _trd,_tsd,_ytr_prices,_yts_prices)

print('\n  Parametri ottimali:')
print(f'  {"Modello":<12} {"units":>6} {"dropout":>8} {"lr":>10} {"wd":>10} {"n_lags":>7} {"MSE_val":>10}')
for mn, hp in best_hp.items():
    print(f'  {mn:<12} {hp["m_units"]:>6} {hp["dropout"]:>8.3f} '
          f'{hp["learn_rate"]:>10.2e} {hp["weight_decay"]:>10.2e} '
          f'{hp["n_lags"]:>7} {hp.get("best_mse_val",float("nan")):>10.6f}')


# ── Helpers globali di valutazione ───────────────────────────
def _smooth_lr(arr, w):
    """Media rolling forward di w passi (costruzione target smoothato)."""
    arr = np.array(arr)
    return np.array([arr[i:i+w].mean() if i+w<=len(arr) else arr[i:].mean()
                     for i in range(len(arr))])

def _raw_pred(model, X_sc):
    """Previsioni inverse-scaled nello spazio log-return."""
    return scaler_y.inverse_transform(predict_pytorch_model(model, X_sc, device)).flatten()

def _metrics(lr_pred, lr_true, lr_smooth=None):
    """RMSE, MSE, MAE, Theil's U, R², DA (su target smoothato se disponibile)."""
    mse  = mean_squared_error(lr_true, lr_pred)
    rmse = np.sqrt(mse)
    mae  = mean_absolute_error(lr_true, lr_pred)
    r2   = r2_score(lr_true, lr_pred)
    tu   = rmse / (np.sqrt(np.mean(lr_true**2))+1e-10)
    ref  = np.sign(lr_smooth if lr_smooth is not None else lr_true)
    da   = np.mean(ref == np.sign(lr_pred)) * 100
    return {'rmse':round(rmse,6),'mse':round(mse,6),'mae':round(mae,6),
            'theil':round(tu,4),'r2':round(r2,4),'da':round(da,2)}

def _prices_from_lr(lr_arr, price_ref):
    return convert_predictions_to_price(lr_arr, price_ref)

# Target smoothato per DA (solo a scopo di valutazione, no leakage)
lr_train_true = scaler_y.inverse_transform(y_train_sc).flatten()
lr_val_true   = scaler_y.inverse_transform(y_val_sc).flatten()
lr_test_true  = scaler_y.inverse_transform(y_test_sc).flatten()
lr_train_sm   = _smooth_lr(lr_train_true, smooth_window) if flag_SMOOTH=='Y' else None
lr_test_sm    = _smooth_lr(lr_test_true,  smooth_window) if flag_SMOOTH=='Y' else None
lr_val_sm     = _smooth_lr(lr_val_true,   smooth_window) if flag_SMOOTH=='Y' else None

y_train_prices= df_btc_op.loc[train_dates_idx,'MKPRU'].values
y_test_prices = df_btc_op.loc[test_dates_idx, 'MKPRU'].values

# Registri globali (popolati nelle celle 7-9)
all_preds_train = {}
all_preds_val   = {}
all_preds_test  = {}
results         = {}
fi_scores_all   = {}
all_lr_val      = {}   # lr_val_true specifico per ogni modello (allineato al suo n_lags)

"""# LSTM Model"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 7 — Modello LSTM                                 ║
# ╚══════════════════════════════════════════════════════════╝

print('='*50)
print('  LSTM')
print('='*50)

hp_lstm = best_hp['LSTM']
(_Xtr_lstm,_Xvl_lstm,_Xts_lstm,_ytr_lstm,_yvl_lstm,_yts_lstm,
 _sy_lstm,sh_lstm,_wf_lstm,_tr1_lstm,_vl1_lstm,_ts1_lstm,
 _lr_tr_lstm,_lr_vl_lstm,_lr_ts_lstm,
 _lr_tr_sm_lstm,_lr_vl_sm_lstm,_lr_ts_sm_lstm,
 _trd_lstm,_tsd_lstm,_ytr_prices_lstm,_yts_prices_lstm) = build_model_split('LSTM')

set_seeds(RND_SEED)
model_lstm = build_lstm_model(sh_lstm, net_units=hp_lstm['m_units'], act_f='tanh',
                               dropout=hp_lstm['dropout'], for_steps=1,
                               opt_rate=hp_lstm['learn_rate'], opt_clip=None, metric='mse')
model_lstm._name = 'LSTM'

print(f'units={hp_lstm["m_units"]} | drop={hp_lstm["dropout"]:.2f} | lr={hp_lstm["learn_rate"]:.1e}')
model_lstm, trl, vll = train_regularized(
    model_lstm, _Xtr_lstm, _ytr_lstm, _Xvl_lstm, _yvl_lstm,
    m_epochs, m_batch_size, hp_lstm['learn_rate'], m_patience,
    device, wd=hp_lstm['weight_decay'], clip=grad_clip,
    dir_alpha=hp_lstm.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

# Walk-forward DA score
if flag_WF=='Y' and _wf_lstm:
    wf_das = []
    for fold in _wf_lstm:
        fp  = _sy_lstm.inverse_transform(
              predict_pytorch_model(model_lstm, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH=='Y' else ft
        wf_das.append(np.mean(np.sign(ref)==np.sign(fp))*100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

# Salva previsioni
all_preds_train['LSTM'] = _sy_lstm.inverse_transform(predict_pytorch_model(model_lstm, _Xtr_lstm, device)).flatten()
all_preds_val['LSTM']   = _sy_lstm.inverse_transform(predict_pytorch_model(model_lstm, _Xvl_lstm, device)).flatten()
all_lr_val['LSTM']      = _lr_vl_lstm
all_preds_test['LSTM']  = _sy_lstm.inverse_transform(predict_pytorch_model(model_lstm, _Xts_lstm, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['LSTM'], _lr_tr_lstm, _lr_tr_sm_lstm)
m_te = _metrics(all_preds_test['LSTM'],  _lr_ts_lstm, _lr_ts_sm_lstm)
results['LSTM'] = {**{f'{k}_train':v for k,v in m_tr.items()},
                   **{f'{k}_test':v  for k,v in m_te.items()}}

# Grafici
fig, axes = plt.subplots(1, 2, figsize=(16,4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes,[
    (all_preds_train['LSTM'], _lr_tr_lstm, _ytr_prices_lstm, _tr1_lstm, _trd_lstm, 'LSTM — Training', _lr_tr_sm_lstm),
    (all_preds_test['LSTM'],  _lr_ts_lstm, _yts_prices_lstm, _ts1_lstm, _tsd_lstm,  'LSTM — Test OOS', _lr_ts_sm_lstm),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Real')
    ax.plot(pp, color='steelblue', linestyle='--', label='Forecast')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

# Feature importance
if flag_FIMP=='Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_lstm, _Xtr_lstm, _ytr_lstm,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['LSTM'] = imp_df.copy()
    plt.figure(figsize=(12,5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='steelblue')
    plt.title('Feature Importance — LSTM (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nLSTM | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'LSTM | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# GRU"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 8 — Modello GRU                                  ║
# ╚══════════════════════════════════════════════════════════╝

print('='*50)
print('  GRU')
print('='*50)

hp_gru = best_hp['GRU']
(_Xtr_gru,_Xvl_gru,_Xts_gru,_ytr_gru,_yvl_gru,_yts_gru,
 _sy_gru,sh_gru,_wf_gru,_tr1_gru,_vl1_gru,_ts1_gru,
 _lr_tr_gru,_lr_vl_gru,_lr_ts_gru,
 _lr_tr_sm_gru,_lr_vl_sm_gru,_lr_ts_sm_gru,
 _trd_gru,_tsd_gru,_ytr_prices_gru,_yts_prices_gru) = build_model_split('GRU')

set_seeds(RND_SEED)
model_gru = build_gru_model(sh_gru, net_units=hp_gru['m_units'], act_f='tanh',
                             dropout=hp_gru['dropout'], for_steps=1,
                             opt_rate=hp_gru['learn_rate'], opt_clip=None, metric='mse')
model_gru._name = 'GRU'

print(f'units={hp_gru["m_units"]} | drop={hp_gru["dropout"]:.2f} | lr={hp_gru["learn_rate"]:.1e}')
model_gru, trl, vll = train_regularized(
    model_gru, _Xtr_gru, _ytr_gru, _Xvl_gru, _yvl_gru,
    m_epochs, m_batch_size, hp_gru['learn_rate'], m_patience,
    device, wd=hp_gru['weight_decay'], clip=grad_clip,
    dir_alpha=hp_gru.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF=='Y' and _wf_gru:
    wf_das = []
    for fold in _wf_gru:
        fp  = _sy_gru.inverse_transform(
              predict_pytorch_model(model_gru, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH=='Y' else ft
        wf_das.append(np.mean(np.sign(ref)==np.sign(fp))*100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['GRU'] = _sy_gru.inverse_transform(predict_pytorch_model(model_gru, _Xtr_gru, device)).flatten()
all_preds_val['GRU']   = _sy_gru.inverse_transform(predict_pytorch_model(model_gru, _Xvl_gru, device)).flatten()
all_lr_val['GRU']      = _lr_vl_gru
all_preds_test['GRU']  = _sy_gru.inverse_transform(predict_pytorch_model(model_gru, _Xts_gru, device)).flatten()

# lr_true già calcolati da build_model_split — nessun ricalcolo
m_tr = _metrics(all_preds_train['GRU'], _lr_tr_gru, _lr_tr_sm_gru)
m_te = _metrics(all_preds_test['GRU'],  _lr_ts_gru, _lr_ts_sm_gru)
results['GRU'] = {**{f'{k}_train':v for k,v in m_tr.items()},
                  **{f'{k}_test':v  for k,v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16,4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes,[
    (all_preds_train['GRU'], _lr_tr_gru, _ytr_prices_gru, _tr1_gru, _trd_gru, 'GRU — Training', _lr_tr_sm_gru),
    (all_preds_test['GRU'],  _lr_ts_gru, _yts_prices_gru, _ts1_gru, _tsd_gru, 'GRU — Test OOS', _lr_ts_sm_gru),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#e67e22', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP=='Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:sh_gru[1]]
    imp_df = compute_feature_importance_lstm(model_gru, _Xtr_gru, _ytr_gru,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['GRU'] = imp_df.copy()
    plt.figure(figsize=(12,5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#e67e22')
    plt.title('Feature Importance — GRU (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nGRU | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'GRU | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""BI-LSTM"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 9 — Modello BI-LSTM                              ║
# ╚══════════════════════════════════════════════════════════╝

print('='*50)
print('  BI-LSTM')
print('='*50)

hp_bi  = best_hp['BI-LSTM']
(_Xtr_bi,_Xvl_bi,_Xts_bi,_ytr_bi,_yvl_bi,_yts_bi,
 _sy_bi,sh_bi,_wf_bi,_tr1_bi,_vl1_bi,_ts1_bi,
 _lr_tr_bi,_lr_vl_bi,_lr_ts_bi,
 _lr_tr_sm_bi,_lr_vl_sm_bi,_lr_ts_sm_bi,
 _trd_bi,_tsd_bi,_ytr_prices_bi,_yts_prices_bi) = build_model_split('BI-LSTM')

set_seeds(RND_SEED)
model_bilstm = build_bilstm_model(sh_bi, net_units=hp_bi['m_units'], act_f='tanh',
                                   dropout=hp_bi['dropout'], for_steps=1,
                                   opt_rate=hp_bi['learn_rate'], opt_clip=None, metric='mse')
model_bilstm._name = 'BI-LSTM'

print(f'units={hp_bi["m_units"]} | drop={hp_bi["dropout"]:.2f} | lr={hp_bi["learn_rate"]:.1e}')
model_bilstm, trl, vll = train_regularized(
    model_bilstm, _Xtr_bi, _ytr_bi, _Xvl_bi, _yvl_bi,
    m_epochs, m_batch_size, hp_bi['learn_rate'], m_patience,
    device, wd=hp_bi['weight_decay'], clip=grad_clip,
    dir_alpha=hp_bi.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF=='Y' and _wf_bi:
    wf_das = []
    for fold in _wf_bi:
        fp  = _sy_bi.inverse_transform(
              predict_pytorch_model(model_bilstm, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH=='Y' else ft
        wf_das.append(np.mean(np.sign(ref)==np.sign(fp))*100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['BI-LSTM'] = _sy_bi.inverse_transform(predict_pytorch_model(model_bilstm, _Xtr_bi, device)).flatten()
all_preds_val['BI-LSTM']   = _sy_bi.inverse_transform(predict_pytorch_model(model_bilstm, _Xvl_bi, device)).flatten()
all_lr_val['BI-LSTM']  = _lr_vl_bi
all_preds_test['BI-LSTM']  = _sy_bi.inverse_transform(predict_pytorch_model(model_bilstm, _Xts_bi, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['BI-LSTM'], _lr_tr_bi, _lr_tr_sm_bi)
m_te = _metrics(all_preds_test['BI-LSTM'],  _lr_ts_bi, _lr_ts_sm_bi)
results['BI-LSTM'] = {**{f'{k}_train':v for k,v in m_tr.items()},
                      **{f'{k}_test':v  for k,v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16,4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes,[
    (all_preds_train['BI-LSTM'], _lr_tr_bi, _ytr_prices_bi, _tr1_bi, _trd_bi, 'BI-LSTM — Training', _lr_tr_sm_bi),
    (all_preds_test['BI-LSTM'],  _lr_ts_bi, _yts_prices_bi, _ts1_bi, _tsd_bi, 'BI-LSTM — Test OOS', _lr_ts_sm_bi),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#8e44ad', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP=='Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_bilstm, _Xtr_bi, _ytr_bi,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['BI-LSTM'] = imp_df.copy()
    plt.figure(figsize=(12,5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#8e44ad')
    plt.title('Feature Importance — BI-LSTM (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nBI-LSTM | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'BI-LSTM | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# LSTM + Attention"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 10-A — Modello LSTM + Attention (Bahdanau)       ║
# ╚══════════════════════════════════════════════════════════╝
# Additive attention (Bahdanau 2015): layer di attention aggiunto
# sopra gli hidden states LSTM. Il modello calcola pesi alfa_t per
# ogni time step della finestra — "guarda indietro" selettivamente
# ai giorni piu rilevanti per la previsione corrente.
# Architettura: Input → LSTM(units) → Attention → Dense(1)
# Usa gli stessi iperparametri Optuna del modello LSTM base.

print('='*50)
print('  LSTM + Attention (Bahdanau)')
print('='*50)

hp_lstm_att = best_hp['LSTM-ATT']
(_Xtr_la,_Xvl_la,_Xts_la,_ytr_la,_yvl_la,_yts_la,
 _sy_la,sh_lstm_att,_wf_la,_tr1_la,_vl1_la,_ts1_la,
 _lr_tr_la,_lr_vl_la,_lr_ts_la,
 _lr_tr_sm_la,_lr_vl_sm_la,_lr_ts_sm_la,
 _trd_la,_tsd_la,_ytr_prices_la,_yts_prices_la) = build_model_split('LSTM-ATT')

set_seeds(RND_SEED)
model_lstm_att = build_lstm_attention_model(
    sh_lstm_att, net_units=hp_lstm_att['m_units'], act_f='tanh',
    dropout=hp_lstm_att['dropout'], for_steps=1,
    opt_rate=hp_lstm_att['learn_rate'], opt_clip=None, metric='mse')
model_lstm_att._name = 'LSTM-ATT'

print(f'units={hp_lstm_att["m_units"]} | drop={hp_lstm_att["dropout"]:.2f} | lr={hp_lstm_att["learn_rate"]:.1e}')
model_lstm_att, trl, vll = train_regularized(
    model_lstm_att, _Xtr_la, _ytr_la, _Xvl_la, _yvl_la,
    m_epochs, m_batch_size, hp_lstm_att['learn_rate'], m_patience,
    device, wd=hp_lstm_att['weight_decay'], clip=grad_clip,
    dir_alpha=hp_lstm_att.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF == 'Y' and _wf_la:
    wf_das = []
    for fold in _wf_la:
        fp  = _sy_la.inverse_transform(
              predict_pytorch_model(model_lstm_att, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH == 'Y' else ft
        wf_das.append(np.mean(np.sign(ref) == np.sign(fp)) * 100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['LSTM-ATT'] = _sy_la.inverse_transform(predict_pytorch_model(model_lstm_att, _Xtr_la, device)).flatten()
all_preds_val['LSTM-ATT']   = _sy_la.inverse_transform(predict_pytorch_model(model_lstm_att, _Xvl_la, device)).flatten()
all_lr_val['LSTM-ATT'] = _lr_vl_la
all_preds_test['LSTM-ATT']  = _sy_la.inverse_transform(predict_pytorch_model(model_lstm_att, _Xts_la, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['LSTM-ATT'], _lr_tr_la, _lr_tr_sm_la)
m_te = _metrics(all_preds_test['LSTM-ATT'],  _lr_ts_la, _lr_ts_sm_la)
results['LSTM-ATT'] = {**{f'{k}_train': v for k, v in m_tr.items()},
                       **{f'{k}_test':  v for k, v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes, [
    (all_preds_train['LSTM-ATT'], _lr_tr_la, _ytr_prices_la, _tr1_la, _trd_la, 'LSTM+Att — Training', _lr_tr_sm_la),
    (all_preds_test['LSTM-ATT'],  _lr_ts_la, _yts_prices_la, _ts1_la, _tsd_la, 'LSTM+Att — Test OOS', _lr_ts_sm_la),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#16a085', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP == 'Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_lstm_att, _Xtr_la, _ytr_la,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['LSTM-ATT'] = imp_df.copy()
    plt.figure(figsize=(12, 5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#16a085')
    plt.title('Feature Importance — LSTM+Attention (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nLSTM-ATT | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'LSTM-ATT | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# GRU + Attention"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 10-B — Modello GRU + Attention (Bahdanau)        ║
# ╚══════════════════════════════════════════════════════════╝
# Stessa logica di LSTM+Att ma con GRU come encoder.
# GRU e piu leggero (2 gate vs 3) e spesso converge piu velocemente.
# Architettura: Input → GRU(units) → Attention → Dense(1)
# Usa gli stessi iperparametri Optuna del modello GRU base.

print('='*50)
print('  GRU + Attention (Bahdanau)')
print('='*50)

hp_gru_att = best_hp['GRU-ATT']
(_Xtr_ga,_Xvl_ga,_Xts_ga,_ytr_ga,_yvl_ga,_yts_ga,
 _sy_ga,sh_gru_att,_wf_ga,_tr1_ga,_vl1_ga,_ts1_ga,
 _lr_tr_ga,_lr_vl_ga,_lr_ts_ga,
 _lr_tr_sm_ga,_lr_vl_sm_ga,_lr_ts_sm_ga,
 _trd_ga,_tsd_ga,_ytr_prices_ga,_yts_prices_ga) = build_model_split('GRU-ATT')

set_seeds(RND_SEED)
model_gru_att = build_gru_attention_model(
    sh_gru_att, net_units=hp_gru_att['m_units'], act_f='tanh',
    dropout=hp_gru_att['dropout'], for_steps=1,
    opt_rate=hp_gru_att['learn_rate'], opt_clip=None, metric='mse')
model_gru_att._name = 'GRU-ATT'

print(f'units={hp_gru_att["m_units"]} | drop={hp_gru_att["dropout"]:.2f} | lr={hp_gru_att["learn_rate"]:.1e}')
model_gru_att, trl, vll = train_regularized(
    model_gru_att, _Xtr_ga, _ytr_ga, _Xvl_ga, _yvl_ga,
    m_epochs, m_batch_size, hp_gru_att['learn_rate'], m_patience,
    device, wd=hp_gru_att['weight_decay'], clip=grad_clip,
    dir_alpha=hp_gru_att.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF == 'Y' and _wf_ga:
    wf_das = []
    for fold in _wf_ga:
        fp  = _sy_ga.inverse_transform(
              predict_pytorch_model(model_gru_att, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH == 'Y' else ft
        wf_das.append(np.mean(np.sign(ref) == np.sign(fp)) * 100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['GRU-ATT'] = _sy_ga.inverse_transform(predict_pytorch_model(model_gru_att, _Xtr_ga, device)).flatten()
all_preds_val['GRU-ATT']   = _sy_ga.inverse_transform(predict_pytorch_model(model_gru_att, _Xvl_ga, device)).flatten()
all_lr_val['GRU-ATT']  = _lr_vl_ga
all_preds_test['GRU-ATT']  = _sy_ga.inverse_transform(predict_pytorch_model(model_gru_att, _Xts_ga, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['GRU-ATT'], _lr_tr_ga, _lr_tr_sm_ga)
m_te = _metrics(all_preds_test['GRU-ATT'],  _lr_ts_ga, _lr_ts_sm_ga)
results['GRU-ATT'] = {**{f'{k}_train': v for k, v in m_tr.items()},
                      **{f'{k}_test':  v for k, v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes, [
    (all_preds_train['GRU-ATT'], _lr_tr_ga, _ytr_prices_ga, _tr1_ga, _trd_ga, 'GRU+Att — Training', _lr_tr_sm_ga),
    (all_preds_test['GRU-ATT'],  _lr_ts_ga, _yts_prices_ga, _ts1_ga, _tsd_ga, 'GRU+Att — Test OOS', _lr_ts_sm_ga),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#d35400', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP == 'Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_gru_att, _Xtr_ga, _ytr_ga,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['GRU-ATT'] = imp_df.copy()
    plt.figure(figsize=(12, 5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#d35400')
    plt.title('Feature Importance — GRU+Attention (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nGRU-ATT | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'GRU-ATT | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# BI-LSTM + Attention"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 10-C — Modello BI-LSTM + Attention (Bahdanau)    ║
# ╚══════════════════════════════════════════════════════════╝
# BI-LSTM processa la sequenza in entrambe le direzioni (forward +
# backward), concatenando i due hidden states. Il layer di attention
# opera sulla sequenza bidirezionale — pesi su 2*units dimensioni.
# Architettura: Input → [LSTMfwd || LSTMbwd] → Attention → Dense(1)
# Usa gli stessi iperparametri Optuna del modello BI-LSTM base.

print('='*50)
print('  BI-LSTM + Attention (Bahdanau)')
print('='*50)

hp_bi_att = best_hp['BI-LSTM-ATT']
(_Xtr_ba,_Xvl_ba,_Xts_ba,_ytr_ba,_yvl_ba,_yts_ba,
 _sy_ba,sh_bi_att,_wf_ba,_tr1_ba,_vl1_ba,_ts1_ba,
 _lr_tr_ba,_lr_vl_ba,_lr_ts_ba,
 _lr_tr_sm_ba,_lr_vl_sm_ba,_lr_ts_sm_ba,
 _trd_ba,_tsd_ba,_ytr_prices_ba,_yts_prices_ba) = build_model_split('BI-LSTM-ATT')

set_seeds(RND_SEED)
model_bi_att = build_bilstm_attention_model(
    sh_bi_att, net_units=hp_bi_att['m_units'], act_f='tanh',
    dropout=hp_bi_att['dropout'], for_steps=1,
    opt_rate=hp_bi_att['learn_rate'], opt_clip=None, metric='mse')
model_bi_att._name = 'BI-LSTM-ATT'

print(f'units={hp_bi_att["m_units"]} | drop={hp_bi_att["dropout"]:.2f} | lr={hp_bi_att["learn_rate"]:.1e}')
model_bi_att, trl, vll = train_regularized(
    model_bi_att, _Xtr_ba, _ytr_ba, _Xvl_ba, _yvl_ba,
    m_epochs, m_batch_size, hp_bi_att['learn_rate'], m_patience,
    device, wd=hp_bi_att['weight_decay'], clip=grad_clip,
    dir_alpha=hp_bi_att.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF == 'Y' and _wf_ba:
    wf_das = []
    for fold in _wf_ba:
        fp  = _sy_ba.inverse_transform(
              predict_pytorch_model(model_bi_att, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH == 'Y' else ft
        wf_das.append(np.mean(np.sign(ref) == np.sign(fp)) * 100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['BI-LSTM-ATT'] = _sy_ba.inverse_transform(predict_pytorch_model(model_bi_att, _Xtr_ba, device)).flatten()
all_preds_val['BI-LSTM-ATT']   = _sy_ba.inverse_transform(predict_pytorch_model(model_bi_att, _Xvl_ba, device)).flatten()
all_lr_val['BI-LSTM-ATT'] = _lr_vl_ba
all_preds_test['BI-LSTM-ATT']  = _sy_ba.inverse_transform(predict_pytorch_model(model_bi_att, _Xts_ba, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['BI-LSTM-ATT'], _lr_tr_ba, _lr_tr_sm_ba)
m_te = _metrics(all_preds_test['BI-LSTM-ATT'],  _lr_ts_ba, _lr_ts_sm_ba)
results['BI-LSTM-ATT'] = {**{f'{k}_train': v for k, v in m_tr.items()},
                          **{f'{k}_test':  v for k, v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes, [
    (all_preds_train['BI-LSTM-ATT'], _lr_tr_ba, _ytr_prices_ba, _tr1_ba, _trd_ba, 'BI-LSTM+Att — Training', _lr_tr_sm_ba),
    (all_preds_test['BI-LSTM-ATT'],  _lr_ts_ba, _yts_prices_ba, _ts1_ba, _tsd_ba, 'BI-LSTM+Att — Test OOS', _lr_ts_sm_ba),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#6c3483', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP == 'Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_bi_att, _Xtr_ba, _ytr_ba,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['BI-LSTM-ATT'] = imp_df.copy()
    plt.figure(figsize=(12, 5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#6c3483')
    plt.title('Feature Importance — BI-LSTM+Attention (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nBI-LSTM-ATT | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'BI-LSTM-ATT | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# Self-Attention (Transformer Like)"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 10-D — Modello Self-Attention (Transformer-like) ║
# ╚══════════════════════════════════════════════════════════╝
# Self-Attention (scaled dot-product, Vaswani et al. 2017):
# ogni time step interagisce con tutti gli altri simultaneamente
# tramite Q=XW^Q, K=XW^K, V=XW^V.
# Attn(Q,K,V) = softmax(QK^T / sqrt(d_k)) * V
# NON usa ricorrenza — cattura dipendenze globali nella finestra
# senza il vincolo di ordinamento sequenziale dell'LSTM/GRU.
# Usa gli stessi iperparametri Optuna del modello LSTM base
# (non c'e un best_hp dedicato per Self-Att).

print('='*50)
print('  Self-Attention (Scaled Dot-Product)')
print('='*50)

hp_self_att = best_hp['SELF-ATT']
(_Xtr_sa,_Xvl_sa,_Xts_sa,_ytr_sa,_yvl_sa,_yts_sa,
 _sy_sa,sh_self_att,_wf_sa,_tr1_sa,_vl1_sa,_ts1_sa,
 _lr_tr_sa,_lr_vl_sa,_lr_ts_sa,
 _lr_tr_sm_sa,_lr_vl_sm_sa,_lr_ts_sm_sa,
 _trd_sa,_tsd_sa,_ytr_prices_sa,_yts_prices_sa) = build_model_split('SELF-ATT')

set_seeds(RND_SEED)
model_self_att = build_self_attention_model(
    sh_self_att, net_units=hp_self_att['m_units'], act_f='tanh',
    dropout=hp_self_att['dropout'], for_steps=1,
    opt_rate=hp_self_att['learn_rate'], opt_clip=None, metric='mse')
model_self_att._name = 'SELF-ATT'

print(f'units={hp_self_att["m_units"]} | drop={hp_self_att["dropout"]:.2f} | lr={hp_self_att["learn_rate"]:.1e}')
model_self_att, trl, vll = train_regularized(
    model_self_att, _Xtr_sa, _ytr_sa, _Xvl_sa, _yvl_sa,
    m_epochs, m_batch_size, hp_self_att['learn_rate'], m_patience,
    device, wd=hp_self_att['weight_decay'], clip=grad_clip,
    dir_alpha=hp_self_att.get('dir_alpha', 0.0))

plot_training_history(trl, vll)

if flag_WF == 'Y' and _wf_sa:
    wf_das = []
    for fold in _wf_sa:
        fp  = _sy_sa.inverse_transform(
              predict_pytorch_model(model_self_att, fold['X_val'], device)).flatten()
        ft  = fold['sc_y'].inverse_transform(fold['y_val']).flatten()
        ref = _smooth_lr(ft, smooth_window) if flag_SMOOTH == 'Y' else ft
        wf_das.append(np.mean(np.sign(ref) == np.sign(fp)) * 100)
    print(f'WF DA: {np.mean(wf_das):.2f}% [{", ".join(f"{d:.1f}" for d in wf_das)}]')

all_preds_train['SELF-ATT'] = _sy_sa.inverse_transform(predict_pytorch_model(model_self_att, _Xtr_sa, device)).flatten()
all_preds_val['SELF-ATT']   = _sy_sa.inverse_transform(predict_pytorch_model(model_self_att, _Xvl_sa, device)).flatten()
all_lr_val['SELF-ATT'] = _lr_vl_sa
all_preds_test['SELF-ATT']  = _sy_sa.inverse_transform(predict_pytorch_model(model_self_att, _Xts_sa, device)).flatten()

# lr_true e smooth già calcolati da build_model_split — nessun ricalcolo necessario
m_tr = _metrics(all_preds_train['SELF-ATT'], _lr_tr_sa, _lr_tr_sm_sa)
m_te = _metrics(all_preds_test['SELF-ATT'],  _lr_ts_sa, _lr_ts_sm_sa)
results['SELF-ATT'] = {**{f'{k}_train': v for k, v in m_tr.items()},
                       **{f'{k}_test':  v for k, v in m_te.items()}}

fig, axes = plt.subplots(1, 2, figsize=(16, 4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes, [
    (all_preds_train['SELF-ATT'], _lr_tr_sa, _ytr_prices_sa, _tr1_sa, _trd_sa, 'Self-Att — Training', _lr_tr_sm_sa),
    (all_preds_test['SELF-ATT'],  _lr_ts_sa, _yts_prices_sa, _ts1_sa, _tsd_sa, 'Self-Att — Test OOS', _lr_ts_sm_sa),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='#c0392b', linestyle='--', label='Previsione')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

if flag_FIMP == 'Y':
    feat_cols = data_model.drop(columns=['MKPRU']).columns.tolist()[:input_shape[1]]
    imp_df = compute_feature_importance_lstm(model_self_att, _Xtr_sa, _ytr_sa,
                                              feat_cols, mean_squared_error,
                                              n_repeats=5, device=device)
    fi_scores_all['SELF-ATT'] = imp_df.copy()
    plt.figure(figsize=(12, 5))
    plt.barh(imp_df['Feature'][:20], imp_df['Importance'][:20], color='#c0392b')
    plt.title('Feature Importance — Self-Attention (Top 20)')
    plt.gca().invert_yaxis(); plt.tight_layout(); plt.show()

print(f'\nSELF-ATT | Train: RMSE={m_tr["rmse"]} R²={m_tr["r2"]} DA={m_tr["da"]}%')
print(f'SELF-ATT | Test:  RMSE={m_te["rmse"]} R²={m_te["r2"]} DA={m_te["da"]}%')

"""# Sommario Modelli"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 10 — Riepilogo individuale                       ║
# ╚══════════════════════════════════════════════════════════╝

results_df = pd.DataFrame(results).T.astype(float)
cols_te    = ['rmse_test','mse_test','mae_test','theil_test','r2_test','da_test']
cols_tr    = ['rmse_train','mse_train','mae_train','theil_train','r2_train','da_train']

print('\n📊 TEST OOS:')
display(results_df[cols_te].style
    .highlight_min(subset=['rmse_test','mse_test','mae_test','theil_test'], color='lightgreen')
    .highlight_max(subset=['r2_test','da_test'], color='lightgreen')
    .format(precision=4))

print('\n📊 TRAINING:')
display(results_df[cols_tr].style
    .highlight_min(subset=['rmse_train','mse_train','mae_train','theil_train'], color='lightgreen')
    .highlight_max(subset=['r2_train','da_train'], color='lightgreen')
    .format(precision=4))

"""# Ensemble con filtro qualità"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 11 — Ensemble adattivo                           ║
# ╚══════════════════════════════════════════════════════════╝
# Filtro qualità su validation set (leak-free).
# Pesi: 1/MSE_val sui modelli qualificati.

print('='*60)
print('  ENSEMBLE — Filtro qualità (val set)')
print(f'  Criteri: R²>{ENS_MIN_R2} | DA>{ENS_MIN_DA}% | Theil<{ENS_MAX_THEIL}')
print('='*60)

qualified = []
ens_mse_val = {}

print(f'\n  {"Modello":<12} {"R²_val":>8} {"DA_val":>8} {"Theil_val":>10} {"Ammesso?":>10}')
print('  ' + '-'*50)

for mname in all_preds_test:
    vp = all_preds_val[mname]
    vt = all_lr_val.get(mname, lr_val_true)
    vt_sm = _smooth_lr(vt, smooth_window) if flag_SMOOTH == 'Y' else None
    r2_v  = float(r2_score(vt, vp))
    ref_v = np.sign(vt_sm if vt_sm is not None else vt)
    da_v  = float(np.mean(ref_v == np.sign(vp))*100)
    rmse_v= float(np.sqrt(mean_squared_error(vt, vp)))
    tu_v  = rmse_v / (np.sqrt(np.mean(vt**2))+1e-10)
    mse_v = float(mean_squared_error(vt, vp))
    ens_mse_val[mname] = mse_v

    ok = r2_v>ENS_MIN_R2 and da_v>ENS_MIN_DA and tu_v<ENS_MAX_THEIL
    reason = []
    if r2_v<=ENS_MIN_R2:    reason.append(f'R²={r2_v:.3f}')
    if da_v<=ENS_MIN_DA:    reason.append(f'DA={da_v:.1f}%')
    if tu_v>=ENS_MAX_THEIL: reason.append(f'U={tu_v:.3f}')
    status = '✅ SI' if ok else f'❌ NO ({", ".join(reason)})'
    print(f'  {mname:<12} {r2_v:>8.4f} {da_v:>7.1f}% {tu_v:>10.4f}  {status}')
    if ok: qualified.append(mname)

if not qualified:
    print('\n⚠️  Nessun modello qualificato — fallback: ensemble completo')
    qualified = list(all_preds_test.keys())

raw_w = {m: 1.0/ens_mse_val[m] for m in qualified}
tot_w = sum(raw_w.values())
ens_w = {m: w/tot_w for m,w in raw_w.items()}
for m in all_preds_test:
    if m not in ens_w: ens_w[m] = 0.0

print('\n  Pesi ensemble:')
for m, w in ens_w.items():
    if w > 0: print(f'  {m:<12} {w*100:.2f}%')

# Previsioni ensemble
# Filtra solo i modelli con peso > 0 — i modelli esclusi hanno peso 0
# ma array di lunghezza diversa (n_lags ≠ consensus) → mismatch di shape.
_active = {m: w for m, w in ens_w.items() if w > 0}
_ens_tr_len = min(len(all_preds_train[m]) for m in _active)
_ens_ts_len = min(len(all_preds_test[m])  for m in _active)
_ens_vl_len = min(len(all_preds_val[m])   for m in _active)
ens_train_lr = sum(w * all_preds_train[m][-_ens_tr_len:] for m, w in _active.items())
ens_test_lr  = sum(w * all_preds_test[m][-_ens_ts_len:]  for m, w in _active.items())
ens_val_lr   = sum(w * all_preds_val[m][-_ens_vl_len:]   for m, w in _active.items())
_lr_tr_ens = lr_train_true[-_ens_tr_len:]
_lr_ts_ens = lr_test_true[-_ens_ts_len:]
_lr_tr_sm_ens = lr_train_sm[-_ens_tr_len:] if lr_train_sm is not None else None
_lr_ts_sm_ens = lr_test_sm[-_ens_ts_len:]  if lr_test_sm  is not None else None
_ytr_prices_ens = y_train_prices[-_ens_tr_len:]
_yts_prices_ens = y_test_prices[-_ens_ts_len:]
_trd_ens = train_dates_idx[-_ens_tr_len:]
_tsd_ens = test_dates_idx[-_ens_ts_len:]

m_ens_tr = _metrics(ens_train_lr, _lr_tr_ens, _lr_tr_sm_ens)
m_ens_te = _metrics(ens_test_lr,  _lr_ts_ens, _lr_ts_sm_ens)
results['ENSEMBLE'] = {**{f'{k}_train':v for k,v in m_ens_tr.items()},
                       **{f'{k}_test':v  for k,v in m_ens_te.items()}}

# Grafici ensemble
fig, axes = plt.subplots(1, 2, figsize=(16,4))
for ax, (lr_p, lr_t, pr_t, pr1, dates, title, sm_ref) in zip(axes,[
    (ens_train_lr, _lr_tr_ens, _ytr_prices_ens, train_price_t1[-_ens_tr_len:], _trd_ens, 'Ensemble — Training', _lr_tr_sm_ens),
    (ens_test_lr,  _lr_ts_ens, _yts_prices_ens, test_price_t1[-_ens_ts_len:],  _tsd_ens, 'Ensemble — Test OOS', _lr_ts_sm_ens),
]):
    pp = pd.Series(_prices_from_lr(lr_p, pr1), index=dates)
    pt = pd.Series(pr_t, index=dates)
    m  = _metrics(lr_p, lr_t, sm_ref)
    ax.plot(pt, color='orange', label='Reale')
    ax.plot(pp, color='crimson', linestyle='--', label='Ensemble')
    ax.set_title(f'{title} | RMSE={m["rmse"]} | Theil={m["theil"]} | DA={m["da"]}%')
    ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout(); plt.show()

print(f'\nEnsemble | Train: RMSE={m_ens_tr["rmse"]} R²={m_ens_tr["r2"]} DA={m_ens_tr["da"]}%')
print(f'Ensemble | Test:  RMSE={m_ens_te["rmse"]} R²={m_ens_te["r2"]} DA={m_ens_te["da"]}%')

"""# Test di robustezza"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 12 — Test di robustezza statistica               ║
# ╚══════════════════════════════════════════════════════════╝
# Test 1: Stabilità temporale (H1 vs H2)
# Test 2: Confronto con benchmark naive
# Test 3: Bootstrap CI sulla DA (1000 resamples)

ref_dir = np.sign(lr_test_sm if lr_test_sm is not None else lr_test_true)
n_obs   = len(ref_dir)
half    = n_obs // 2
all_preds_eval = {**all_preds_test, 'ENSEMBLE': ens_test_lr}

print(f'\n  {n_obs} campioni di test | rischio fortuna: {"MODERATO" if n_obs<100 else "BASSO"}')

# ── Test 1 — Stabilità temporale ─────────────────────────────
print(f'\n  TEST 1 — Stabilità temporale')
print(f'  {"Modello":<12} {"DA":>8} {"H1":>8} {"H2":>8} {"Stabile?":>10}')
print('  ' + '-'*44)
stab = {}
for mn, p in all_preds_eval.items():
    pd_ = np.sign(p)
    da, h1, h2 = (np.mean(ref_dir==pd_)*100,
                  np.mean(ref_dir[:half]==pd_[:half])*100,
                  np.mean(ref_dir[half:]==pd_[half:])*100)
    ok = 'SI ✅' if h1>50 and h2>50 else 'NO ❌'
    stab[mn] = ok
    print(f'  {mn:<12} {da:>7.1f}% {h1:>7.1f}% {h2:>7.1f}%  {ok}')

# ── Test 2 — Benchmark naive ──────────────────────────────────
naive_up  = np.mean(ref_dir==1)*100
naive_mom = np.mean(ref_dir==np.sign(np.roll(lr_test_true,1)))*100
best_naive= max(naive_up, naive_mom)

print(f'\n  TEST 2 — Naive benchmark (best naive={best_naive:.1f}%)')
print(f'  {"Modello":<12} {"DA":>8} {"vs naive":>10} {"Batte?":>8}')
print('  ' + '-'*42)
beats = {}
for mn, p in all_preds_eval.items():
    da   = np.mean(ref_dir==np.sign(p))*100
    diff = da - best_naive
    ok   = 'SI ✅' if da > best_naive else 'NO ❌'
    beats[mn] = ok
    print(f'  {mn:<12} {da:>7.1f}% {diff:>+9.1f}%  {ok}')

# ── Test 3 — Bootstrap CI + binomiale ────────────────────────
print(f'\n  TEST 3 — Bootstrap CI (1000 resamples, H0: DA=50%)')
print(f'  {"Modello":<12} {"DA":>8} {"CI_low":>9} {"CI_high":>9} {"p-val":>8} {"Sig?":>6}')
print('  ' + '-'*56)
np.random.seed(42)
boot_res = {}
for mn, p in all_preds_eval.items():
    correct  = (ref_dir==np.sign(p)).astype(float)
    da_obs   = correct.mean()*100
    boot_das = np.array([correct[np.random.randint(0,n_obs,n_obs)].mean()*100
                         for _ in range(1000)])
    ci_lo, ci_hi = np.percentile(boot_das, [2.5, 97.5])
    pval = _stats.binomtest(int(correct.sum()), n_obs, p=0.5, alternative='greater').pvalue
    sig  = 'SI ✅' if pval<0.05 else 'NO ❌'
    boot_res[mn] = sig
    print(f'  {mn:<12} {da_obs:>7.1f}% {ci_lo:>8.1f}% {ci_hi:>8.1f}%  {pval:>7.4f}  {sig}')

# ── Riepilogo ─────────────────────────────────────────────────
print(f'\n  {"="*55}')
print(f'  ROBUSTEZZA — modelli che passano tutti e 3 i test:')
print(f'  {"="*55}')
for mn in all_preds_eval:
    t1 = stab.get(mn,'NO ❌')=='SI ✅'
    t2 = beats.get(mn,'NO ❌')=='SI ✅'
    t3 = boot_res.get(mn,'NO ❌')=='SI ✅'
    status = '✅ ROBUSTO' if (t1 and t2 and t3) else '⚠️  parziale' if (t1 or t2 or t3) else '❌ non robusto'
    print(f'  {mn:<12} T1={("✅" if t1 else "❌")}  T2={("✅" if t2 else "❌")}  T3={("✅" if t3 else "❌")}  → {status}')

"""# Trading Systems"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 14 — Trading Strategy per Model                    ║
# ║  Strategy A (Long/Flat) and B (Long/Short) vs Buy & Hold ║
# ║                                                          ║
# ║  INDEPENDENT CELL — only requires that the model cells   ║
# ║  (7–10D) have been executed to populate all_preds_test,  ║
# ║  lr_test_true, y_test_prices, test_dates_idx.            ║
# ╚══════════════════════════════════════════════════════════╝

# ════════════════════════════════════════════════════════════
# PARAMETERS — modify here before running
# ════════════════════════════════════════════════════════════

TS_TAKE_PR     =  0.0005    # long entry threshold:  pred > +0.05%
TS_STOP_L      = -0.0005    # short/flat threshold:  pred < -0.05%
TS_TRANS_COST  =  0.001     # transaction cost:      0.1%
TS_INIT_INV    =  100_000   # initial capital (USD)

# Models to analyze — remove those not trained
TS_MODELS = [
    'LSTM',
    'GRU',
    'BI-LSTM',
    'LSTM-ATT',
    'GRU-ATT',
    'BI-LSTM-ATT',
    'SELF-ATT',
]

# ════════════════════════════════════════════════════════════
# FUNCTIONS (self-contained, no external dependencies)
# ════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from sklearn.metrics import mean_squared_error

def _ts_run(pred_lr, actual_lr, dates, init_inv,
            trans_cost, take_pr, stop_l, allow_short):
    """
    Simulates the strategy on the test set.

    Strategy A (allow_short=False): Long / Flat
    Strategy B (allow_short=True):  Long / Short / Flat

    Entry rules:
      pred > take_pr  → Long
      pred < stop_l   → Short (B) or Flat (A)
      otherwise       → keep previous position (dead zone)

    Returns a dict with portfolio, metrics, and trade statistics.
    """
    n       = len(pred_lr)
    port    = np.zeros(n + 1);  port[0] = init_inv
    pos     = 0        # 0=flat, 1=long, -1=short
    entry_v = init_inv
    trades  = []
    daily_r = np.zeros(n)

    for t in range(n):
        ret_t = np.exp(actual_lr[t]) - 1

        # New position
        if pred_lr[t] > take_pr:
            new_pos = 1
        elif pred_lr[t] < stop_l:
            new_pos = -1 if allow_short else 0
        else:
            new_pos = pos                      # dead zone

        tc = trans_cost if new_pos != pos else 0.0

        # Close previous position if it changes
        if new_pos != pos:
            if pos != 0:
                trades.append((port[t] * (1-tc) - entry_v) / entry_v)
            entry_v = port[t] * (1 - tc)

        # Daily P&L
        if new_pos == 1:
            port[t+1] = port[t] * (1 - tc) * (1 + ret_t)
        elif new_pos == -1:
            port[t+1] = port[t] * (1 - tc) * (1 - ret_t)
        else:
            port[t+1] = port[t] * (1 - tc)

        daily_r[t] = (port[t+1] - port[t]) / port[t]
        pos = new_pos

        # Close open position at end of period
    if pos != 0:
        trades.append((port[n] - entry_v) / entry_v)

    # ── Metrics ───────────────────────────────────────────────
    total_ret = (port[n] - port[0]) / port[0] * 100
    n_days    = len(actual_lr)
    cagr      = ((port[n] / port[0]) ** (365 / max(n_days, 1)) - 1) * 100
    port_s    = pd.Series(port[1:], index=dates)
    dd        = (port_s - port_s.cummax()) / port_s.cummax() * 100
    max_dd    = dd.min()
    calmar    = cagr / abs(max_dd) if max_dd != 0 else 0.0

    # Sharpe and Sortino on invested days (excludes flat with ret=0)
    inv_r  = daily_r[daily_r != 0]
    sharpe = (inv_r.mean() / (inv_r.std() + 1e-10)) * np.sqrt(252) if len(inv_r) > 1 else 0.0
    neg_r  = inv_r[inv_r < 0]
    sortino= (inv_r.mean() / (neg_r.std() + 1e-10)) * np.sqrt(252) if len(neg_r) > 1 else 0.0

    # Trade stats
    n_tr   = len(trades)
    wins   = [t for t in trades if t > 0]
    losses = [t for t in trades if t <= 0]
    wr     = len(wins) / n_tr * 100 if n_tr > 0 else 0.0
    pf     = sum(wins) / abs(sum(losses)) if losses and sum(losses) != 0 else np.inf
    avg_tr = float(np.mean(trades)) * 100 if trades else 0.0

    # Bootstrap CI Sharpe (1000 resamples on invested days)
    ci_lo, ci_hi = np.nan, np.nan
    if len(inv_r) > 5:
        np.random.seed(42)
        boot = np.array([
            (lambda x: (x.mean() / (x.std() + 1e-10)) * np.sqrt(252))(
                np.random.choice(inv_r, len(inv_r), replace=True))
            for _ in range(1000)
        ])
        ci_lo, ci_hi = np.percentile(boot, [2.5, 97.5])

    # Count days per position
    positions = []
    cur = 0
    for p in pred_lr:
        if p > take_pr:    positions.append(1)
        elif p < stop_l:   positions.append(-1 if allow_short else 0)
        else:              positions.append(cur)
        cur = positions[-1]
    pos_arr = np.array(positions)

    return {
        'port':      port,           'dd':        dd,
        'daily_r':   daily_r,        'total_ret': total_ret,
        'cagr':      cagr,           'sharpe':    sharpe,
        'sortino':   sortino,        'max_dd':    max_dd,
        'calmar':    calmar,         'n_trades':  n_tr,
        'win_rate':  wr,             'pf':        pf,
        'avg_trade': avg_tr,         'ci_lo':     ci_lo,
        'ci_hi':     ci_hi,
        'n_long':    int((pos_arr == 1).sum()),
        'n_short':   int((pos_arr == -1).sum()),
        'n_flat':    int((pos_arr == 0).sum()),
    }


def _bh(actual_lr, dates, init_inv):
    """Buy & Hold baseline."""
    port   = init_inv * np.cumprod(np.exp(actual_lr))
    port_s = pd.Series(port, index=dates)
    dd     = (port_s - port_s.cummax()) / port_s.cummax() * 100
    dr     = np.exp(actual_lr) - 1
    total  = (port[-1] - init_inv) / init_inv * 100
    cagr   = ((port[-1] / init_inv) ** (365 / max(len(actual_lr), 1)) - 1) * 100
    sr     = (dr.mean() / (dr.std() + 1e-10)) * np.sqrt(252)
    so     = (dr.mean() / (dr[dr < 0].std() + 1e-10)) * np.sqrt(252)
    max_dd = dd.min()
    return {
        'port':      np.concatenate([[init_inv], port]),
        'dd':        dd,             'total_ret': total,
        'cagr':      cagr,           'sharpe':    sr,
        'sortino':   so,             'max_dd':    max_dd,
        'calmar':    cagr / abs(max_dd) if max_dd != 0 else 0.0,
    }


def _print_row(label, vA, vB, vBH, w=(22, 14, 14, 12)):
    print(f'  {label:<{w[0]}} {vA:>{w[1]}} {vB:>{w[2]}} {vBH:>{w[3]}}')

def _fmt_pf(v):
    return f'{v:>14.3f}' if v != np.inf else '            ∞'

def _fmt_ci(lo, hi):
    if np.isnan(lo): return 'N/A'
    return f'[{lo:+.3f},{hi:+.3f}]'

def _sig(lo):
    if np.isnan(lo): return '—'
    return '✅' if lo > 0 else '⚠️ '


# ════════════════════════════════════════════════════════════
# EXECUTION
# ════════════════════════════════════════════════════════════

# Verify required variables exist
_required = ['all_preds_test', 'lr_test_true', 'y_test_prices', 'test_dates_idx']
_missing  = [v for v in _required if v not in dir()]
if _missing:
    raise RuntimeError(
        f'Missing variables: {_missing}.\n'
        'Run the model cells (7–10D) first to populate all_preds_test.')

# Filter only the models actually trained
TS_MODELS_OK = [m for m in TS_MODELS if m in all_preds_test]
if len(TS_MODELS_OK) < len(TS_MODELS):
    skipped = set(TS_MODELS) - set(TS_MODELS_OK)
    print(f'⚠️  Models not found in all_preds_test (skipping): {skipped}')

# Buy & Hold
bh = _bh(lr_test_true, test_dates_idx, TS_INIT_INV)

# Simulate A and B for each model
res_A, res_B = {}, {}
for mname in TS_MODELS_OK:
    pred = all_preds_test[mname]
    res_A[mname] = _ts_run(pred, lr_test_true, test_dates_idx,
                            TS_INIT_INV, TS_TRANS_COST,
                            TS_TAKE_PR, TS_STOP_L, allow_short=False)
    res_B[mname] = _ts_run(pred, lr_test_true, test_dates_idx,
                            TS_INIT_INV, TS_TRANS_COST,
                            TS_TAKE_PR, TS_STOP_L, allow_short=True)

print(f'✅ Simulation completed for {len(TS_MODELS_OK)} models')
print(f'   Parameters: TC={TS_TRANS_COST*100:.2f}% | take_pr={TS_TAKE_PR:+.4f} | stop_l={TS_STOP_L:+.4f}')
print(f'   B&H: {bh["total_ret"]:+.2f}% | Sharpe={bh["sharpe"]:.3f} | MaxDD={bh["max_dd"]:.2f}%')


# ════════════════════════════════════════════════════════════
# PLOTS — one per model
# ════════════════════════════════════════════════════════════

# Prepend first day to align portfolio (n+1) to dates (n)
dates_ext = pd.DatetimeIndex([test_dates_idx[0]]).append(test_dates_idx)

for mname in TS_MODELS_OK:
    rA   = res_A[mname]
    rB   = res_B[mname]
    pred = all_preds_test[mname]

    fig, axes = plt.subplots(1, 3, figsize=(20, 4))
    fig.suptitle(
        f'{mname} — Trading Strategy | OOS Test '
        f'({str(test_dates_idx[0].date())} → {str(test_dates_idx[-1].date())})',
        fontsize=12, fontweight='bold')

    # Equity curve
    ax = axes[0]
    ax.plot(dates_ext, rA['port'], color='steelblue', lw=1.8, label='A: Long/Flat')
    ax.plot(dates_ext, rB['port'], color='#e67e22',   lw=1.8, label='B: Long/Short', ls='--')
    ax.plot(dates_ext, bh['port'], color='gray',      lw=1.0, label='Buy & Hold',    ls=':')
    ax.axhline(TS_INIT_INV, color='black', lw=0.5, ls='--', alpha=0.4)
    ax.set_title('Equity Curve ($)')
    ax.legend(fontsize=8); ax.grid(alpha=0.3)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

    # Drawdown
    ax = axes[1]
    ax.fill_between(test_dates_idx, rA['dd'], 0, alpha=0.45, color='steelblue', label='A')
    ax.fill_between(test_dates_idx, rB['dd'], 0, alpha=0.45, color='#e67e22',   label='B')
    ax.fill_between(test_dates_idx, bh['dd'], 0, alpha=0.20, color='gray',      label='B&H')
    ax.set_title('Drawdown (%)'); ax.legend(fontsize=8); ax.grid(alpha=0.3)

    # Price signals
    ax = axes[2]
    colors_sig = [
        '#27ae60' if p > TS_TAKE_PR else
        '#e74c3c' if p < TS_STOP_L  else
        '#bdc3c7'
        for p in pred
    ]
    ax.plot(test_dates_idx, y_test_prices, color='black', lw=0.6, alpha=0.4, zorder=1)
    ax.scatter(test_dates_idx, y_test_prices, c=colors_sig, s=14, zorder=2)
    ax.legend(handles=[
        Patch(color='#27ae60', label=f'Long  (pred>{TS_TAKE_PR:+.4f})'),
        Patch(color='#e74c3c', label=f'Short/Flat (pred<{TS_STOP_L:+.4f})'),
        Patch(color='#bdc3c7', label='Dead zone'),
    ], fontsize=7, loc='upper left')
    ax.set_title('Signals on BTC Price'); ax.grid(alpha=0.3)

    plt.tight_layout(); plt.show()

    # Metrics table
    SEP = '  ' + '-'*66
    HDR = f'  {"Metric":<22} {"A: Long/Flat":>14} {"B: Long/Short":>14} {"Buy & Hold":>12}'
    print(f'\n  ── {mname} ──')
    print(HDR); print(SEP)
    _print_row('Final Value ($)',
               f'${rA["port"][-1]:>11,.0f}', f'${rB["port"][-1]:>11,.0f}',
               f'${bh["port"][-1]:>9,.0f}')
    _print_row('Total Return (%)',
               f'{rA["total_ret"]:>+13.2f}%', f'{rB["total_ret"]:>+13.2f}%',
               f'{bh["total_ret"]:>+11.2f}%')
    _print_row('CAGR (%) [inflated]',
               f'{rA["cagr"]:>+13.2f}%', f'{rB["cagr"]:>+13.2f}%',
               f'{bh["cagr"]:>+11.2f}%')
    _print_row('Sharpe [²]',
               f'{rA["sharpe"]:>14.3f}', f'{rB["sharpe"]:>14.3f}',
               f'{bh["sharpe"]:>12.3f}')
    _print_row('Sortino',
               f'{rA["sortino"]:>14.3f}', f'{rB["sortino"]:>14.3f}',
               f'{bh["sortino"]:>12.3f}')
    _print_row('Max Drawdown (%)',
               f'{rA["max_dd"]:>+13.2f}%', f'{rB["max_dd"]:>+13.2f}%',
               f'{bh["max_dd"]:>+11.2f}%')
    _print_row('Calmar',
               f'{rA["calmar"]:>14.3f}', f'{rB["calmar"]:>14.3f}',
               f'{bh["calmar"]:>12.3f}')
    print(SEP)
    _print_row('N. Trades',
               f'{rA["n_trades"]:>14d}', f'{rB["n_trades"]:>14d}', '            —')
    _print_row('Win Rate (%)',
               f'{rA["win_rate"]:>13.1f}%', f'{rB["win_rate"]:>13.1f}%', '            —')
    _print_row('Profit Factor',
               _fmt_pf(rA["pf"]), _fmt_pf(rB["pf"]), '            —')
    _print_row('Avg Trade (%)',
               f'{rA["avg_trade"]:>+13.2f}%', f'{rB["avg_trade"]:>+13.2f}%', '            —')
    print(SEP)
    _print_row('Long days',
               f'{rA["n_long"]:>14d}', f'{rB["n_long"]:>14d}',
               f'{len(lr_test_true):>12d}')
    _print_row('Short days',
               f'{rA["n_short"]:>14d}', f'{rB["n_short"]:>14d}', '            —')
    _print_row('Flat days',
               f'{rA["n_flat"]:>14d}', f'{rB["n_flat"]:>14d}', '            0')
    print(SEP)
    _print_row('CI Sharpe 95% [boot]',
               _fmt_ci(rA["ci_lo"], rA["ci_hi"]) + ' ' + _sig(rA["ci_lo"]),
               _fmt_ci(rB["ci_lo"], rB["ci_hi"]) + ' ' + _sig(rB["ci_lo"]),
               '            —')
    print()


# ════════════════════════════════════════════════════════════
# COMPARATIVE TABLE — all models
# ════════════════════════════════════════════════════════════

for strat_label, res_dict in [('A — Long/Flat', res_A), ('B — Long/Short', res_B)]:
    print('\n' + '='*85)
    print(f'  SUMMARY — Strategy {strat_label} vs Buy & Hold')
    print('='*85)
    print(f'  {"Model":<14} {"TotRet%":>9} {"Sharpe":>8} {"MaxDD%":>9} '
          f'{"N.Tr":>6} {"WinR%":>7} {"PF":>7} {"AvgTr%":>7}  {"CI Sharpe 95%":>18}')
    print('  ' + '-'*85)
    for mname in TS_MODELS_OK:
        r   = res_dict[mname]
        ci  = _fmt_ci(r["ci_lo"], r["ci_hi"])
        sig = _sig(r["ci_lo"])
        pf  = f'{r["pf"]:>7.2f}' if r["pf"] != np.inf else '      ∞'
        print(f'  {mname:<14} {r["total_ret"]:>+8.2f}% {r["sharpe"]:>8.3f} '
              f'{r["max_dd"]:>+8.2f}% {r["n_trades"]:>6d} {r["win_rate"]:>6.1f}% '
              f'{pf} {r["avg_trade"]:>+6.2f}%  {ci:>18} {sig}')
    print('  ' + '-'*85)
    print(f'  {"Buy & Hold":<14} {bh["total_ret"]:>+8.2f}% {bh["sharpe"]:>8.3f} '
          f'{bh["max_dd"]:>+8.2f}%      —       —       —        —'
          f'                   —')

print('\n  [¹] CAGR is amplified by short windows — do not use as primary metric.')
print('  [²] Sharpe is inflated by flat days with zero return.')
print('  Primary metric: Total Return and Bootstrap CI Sharpe.')
print(f'\n  OOS Test: {test_dates_idx[0].date()} → {test_dates_idx[-1].date()} '
      f'({len(lr_test_true)} days | '
      f'{(test_dates_idx[-1]-test_dates_idx[0]).days/365:.2f} years)')

"""# Advanced Sharpe Ratio"""

# ╔══════════════════════════════════════════════════════════╗
# ║  CELLA 15 — Advanced Sharpe Ratio Analysis              ║
# ║  PSR / DSR / MinTRL / Bootstrap CI (iid + block)        ║
# ║                                                          ║
# ║  Ref: Bailey & Lopez de Prado (2012, 2014)              ║
# ║                                                          ║
# ║  CELLA INDIPENDENTE — richiede solo:                     ║
# ║    all_preds_test, lr_test_true, lr_test_sm,             ║
# ║    test_dates_idx, scaler_y, predict_pytorch_model,      ║
# ║    device (popolati dalle celle modello 7–10D)           ║
# ╚══════════════════════════════════════════════════════════╝

# ════════════════════════════════════════════════════════════
# PARAMETRI — modifica qui
# ════════════════════════════════════════════════════════════

AS_SR_STAR    = 0.0    # benchmark SR* per PSR (0 = batte il caso)
AS_ALPHA      = 0.05   # livello di significatività (5%)
AS_K_TRIALS   = 61     # K configurazioni testate (60 Optuna + 1 ensemble)
                       # usato per DSR — aumenta se hai fatto più trial
AS_BLOCK_SIZE = None   # block_size per block bootstrap
                       # None → usa floor(sqrt(T)) automaticamente

# Modelli da analizzare (deve coincidere con all_preds_test)
AS_MODELS = [
    'LSTM', 'GRU', 'BI-LSTM',
    'LSTM-ATT', 'GRU-ATT', 'BI-LSTM-ATT', 'SELF-ATT',
]

# ════════════════════════════════════════════════════════════
# LIBRERIE
# ════════════════════════════════════════════════════════════

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as _stats
from scipy.special import ndtr  # CDF normale standard (più precisa di norm.cdf)

# ════════════════════════════════════════════════════════════
# FUNZIONI (Bailey & Lopez de Prado)
# ════════════════════════════════════════════════════════════

def _sr_stats(daily_ret, ann_factor=252):
    """
    Calcola SR annualizzato e le sue statistiche descrittive
    sui soli giorni investiti (esclude flat con ret=0).

    Ritorna: sr, mu, sigma, skew (γ3), kurt_pearson (γ4), T
    """
    r  = np.array(daily_ret)
    r  = r[r != 0]           # escludi giorni flat
    T  = len(r)
    if T < 4:
        return np.nan, np.nan, np.nan, np.nan, np.nan, T
    mu    = r.mean()
    sigma = r.std(ddof=1)
    if sigma < 1e-12:
        return np.nan, mu, sigma, np.nan, np.nan, T
    sr    = (mu / sigma) * np.sqrt(ann_factor)
    skew  = float(_stats.skew(r))
    kurt  = float(_stats.kurtosis(r, fisher=False))  # Pearson (normale=3)
    return sr, mu, sigma, skew, kurt, T


def _sr_variance(sr_hat, skew, kurt_p, T):
    """
    Varianza asintotica dello Sharpe ratio stimato (Mertens 2002,
    corretta per skewness e kurtosis di Pearson):

        Var(SR^) ≈ (1/T) * [1 + SR^²/2 - γ3·SR^ + (γ4-1)/4·SR^²]

    dove γ4 è la kurtosis di Pearson (normale = 3).
    """
    sr2  = sr_hat ** 2
    var  = (1 / T) * (1 + sr2/2 - skew*sr_hat + (kurt_p - 1)/4 * sr2)
    return max(var, 1e-12)   # garantisce positività numerica


def psr(sr_hat, sr_star, skew, kurt_p, T):
    """
    Probabilistic Sharpe Ratio (Bailey & Lopez de Prado 2012).
    PSR(SR*) = P(SR_true > SR*) = Φ[(SR^ - SR*) / σ(SR^)]

    sr_star : benchmark (0 = batte il caso, o SR del B&H)
    Ritorna: PSR in [0,1]
    """
    if np.isnan(sr_hat): return np.nan
    var_sr = _sr_variance(sr_hat, skew, kurt_p, T)
    z      = (sr_hat - sr_star) / np.sqrt(var_sr)
    return float(ndtr(z))


def dsr(sr_hat, skew, kurt_p, T, K, alpha=0.05):
    """
    Deflated Sharpe Ratio (Bailey & Lopez de Prado 2014).
    Corregge il PSR per la selezione tra K configurazioni testate.

    Il benchmark atteso sotto K prove iid è approssimato tramite
    Extreme Value Theory (formula di Bailey & Lopez de Prado):

        SR*_K ≈ σ_K · [(1-γ)·Φ⁻¹(1 - 1/K) + γ·Φ⁻¹(1 - 1/(K·e))]

    dove γ = costante di Euler-Mascheroni ≈ 0.5772, e = 2.718...
    σ_K = std delle SR across trials (approssimata con σ(SR^))

    Ritorna: (DSR, SR*_adjusted)
    """
    if np.isnan(sr_hat) or K <= 1: return np.nan, np.nan
    var_sr  = _sr_variance(sr_hat, skew, kurt_p, T)
    sigma_K = np.sqrt(var_sr)           # approssimazione σ across trials
    gamma_e = 0.5772156649             # costante di Euler-Mascheroni
    e       = np.e
    if K == 1:
        sr_adj = 0.0
    else:
        from scipy.stats import norm
        term1   = (1 - gamma_e) * norm.ppf(1 - 1/K)
        term2   = gamma_e       * norm.ppf(1 - 1/(K * e))
        sr_adj  = sigma_K * (term1 + term2)
    dsr_val = psr(sr_hat, sr_adj, skew, kurt_p, T)
    return float(dsr_val), float(sr_adj)


def min_trl(sr_hat, skew, kurt_p, sr_star=0.0, alpha=0.05):
    """
    Minimum Track Record Length (Bailey & Lopez de Prado 2014).
    Numero minimo di osservazioni (giorni investiti) per concludere
    SR_true > SR* con confidenza (1-α):

        MinTRL = 1 + (1 + SR^²/2 - γ3·SR^ + (γ4-1)/4·SR^²)
                     · [z_{1-α} / (SR^ - SR*)]²

    Ritorna MinTRL in giorni investiti (non calendario).
    """
    if np.isnan(sr_hat) or sr_hat <= sr_star: return np.inf
    from scipy.stats import norm
    z      = norm.ppf(1 - alpha)
    factor = 1 + sr_hat**2/2 - skew*sr_hat + (kurt_p-1)/4*sr_hat**2
    trl    = 1 + factor * (z / (sr_hat - sr_star))**2
    return float(trl)


def bootstrap_sr_iid(daily_ret, n_boot=1000, alpha=0.05, ann=252, seed=42):
    """Bootstrap iid: ricampiona giorni investiti con replacement."""
    r = np.array(daily_ret); r = r[r != 0]
    if len(r) < 5: return np.nan, np.nan
    np.random.seed(seed)
    boot = np.array([
        (lambda x: (x.mean()/(x.std(ddof=1)+1e-12))*np.sqrt(ann))(
            np.random.choice(r, len(r), replace=True))
        for _ in range(n_boot)
    ])
    lo, hi = np.percentile(boot, [alpha/2*100, (1-alpha/2)*100])
    return float(lo), float(hi)


def bootstrap_sr_block(daily_ret, block_size=None,
                        n_boot=1000, alpha=0.05, ann=252, seed=42):
    """
    Block bootstrap (Künsch 1989): preserva l'autocorrelazione
    temporale ricampionando blocchi contigui di lunghezza block_size.
    block_size default = floor(√T) dei giorni investiti.
    """
    r = np.array(daily_ret); r = r[r != 0]
    T = len(r)
    if T < 5: return np.nan, np.nan
    bs = block_size if block_size else max(2, int(np.floor(np.sqrt(T))))
    np.random.seed(seed + 1)
    boot = []
    for _ in range(n_boot):
        indices = []
        while len(indices) < T:
            start = np.random.randint(0, T - bs + 1)
            indices.extend(range(start, min(start + bs, T)))
        block_r = r[np.array(indices[:T])]
        sr_b    = (block_r.mean()/(block_r.std(ddof=1)+1e-12))*np.sqrt(ann)
        boot.append(sr_b)
    boot = np.array(boot)
    lo, hi = np.percentile(boot, [alpha/2*100, (1-alpha/2)*100])
    return float(lo), float(hi)


# ════════════════════════════════════════════════════════════
# VERIFICA DIPENDENZE
# ════════════════════════════════════════════════════════════

_required = ['all_preds_test', 'lr_test_true', 'test_dates_idx']
_missing  = [v for v in _required if v not in dir()]
if _missing:
    raise RuntimeError(
        f'Variabili mancanti: {_missing}.\n'
        'Esegui prima le celle modello (7–10D).')

AS_MODELS_OK = [m for m in AS_MODELS if m in all_preds_test]
if len(AS_MODELS_OK) < len(AS_MODELS):
    print(f'⚠️  Skip modelli non trovati: {set(AS_MODELS)-set(AS_MODELS_OK)}')

T_cal      = len(lr_test_true)              # giorni calendario test
bs_default = max(2, int(np.floor(np.sqrt(T_cal))))
block_size = AS_BLOCK_SIZE if AS_BLOCK_SIZE else bs_default

# Smooth target per DA (se disponibile)
_sm = lr_test_sm if 'lr_test_sm' in dir() and lr_test_sm is not None else lr_test_true

print(f'Parametri: SR*={AS_SR_STAR} | α={AS_ALPHA} | K={AS_K_TRIALS} | '
      f'block_size={block_size}d (=√{T_cal}≈{np.sqrt(T_cal):.1f})')
print(f'T calendario={T_cal} | Modelli: {AS_MODELS_OK}')


# ════════════════════════════════════════════════════════════
# CALCOLO — ogni modello
# ════════════════════════════════════════════════════════════

# SR Buy & Hold (giorni calendario, nessun giorno flat)
bh_ret   = np.exp(lr_test_true) - 1
bh_sr    = float((bh_ret.mean()/(bh_ret.std(ddof=1)+1e-12))*np.sqrt(252))

records = []

for mname in AS_MODELS_OK:
    pred     = all_preds_test[mname]

    # Ricostruisci daily returns della Strategy A (Long/Flat)
    # usando le stesse regole della cella 14 con parametri default
    # (prende TS_TAKE_PR/TS_STOP_L se definiti, altrimenti 0.0005)
    _take  = TS_TAKE_PR if 'TS_TAKE_PR' in dir() else 0.0005
    _stop  = TS_STOP_L  if 'TS_STOP_L'  in dir() else -0.0005

    port   = 100_000.0
    pos    = 0
    dr     = []
    for t in range(T_cal):
        ret_t = np.exp(lr_test_true[t]) - 1
        if pred[t] > _take:   new_pos = 1
        elif pred[t] < _stop: new_pos = 0
        else:                  new_pos = pos
        tc = 0.001 if new_pos != pos else 0.0
        if new_pos == 1:
            new_port = port * (1 - tc) * (1 + ret_t)
        else:
            new_port = port * (1 - tc)
        dr.append((new_port - port) / port)
        port = new_port
        pos  = new_pos

    dr_arr   = np.array(dr)
    inv_r    = dr_arr[dr_arr != 0]     # solo giorni investiti
    T_active = len(inv_r)              # T per PSR/DSR/MinTRL

    # SR e statistiche
    sr, mu, sigma, skew, kurt_p, _ = _sr_stats(dr_arr)

    # Jarque-Bera e Ljung-Box (diagnostica)
    jb_p = float(_stats.jarque_bera(inv_r).pvalue) if T_active > 7 else np.nan
    if T_active > 12:
        try:
            from statsmodels.stats.diagnostic import acorr_ljungbox
            lb_res = acorr_ljungbox(inv_r, lags=10, return_df=True)
            lb_p   = float(lb_res['lb_pvalue'].iloc[-1])
        except Exception:
            ac1  = float(np.corrcoef(inv_r[:-1], inv_r[1:])[0,1])
            lb_p = float(_stats.norm.sf(abs(ac1)*np.sqrt(T_active)))
    else:
        lb_p = np.nan

    # PSR
    psr_0  = psr(sr, AS_SR_STAR, skew, kurt_p, T_active)
    psr_bh = psr(sr, bh_sr,      skew, kurt_p, T_active)

    # DSR
    dsr_v, sr_adj = dsr(sr, skew, kurt_p, T_active, AS_K_TRIALS, AS_ALPHA)

    # MinTRL
    mtrl = min_trl(sr, skew, kurt_p, sr_star=AS_SR_STAR, alpha=AS_ALPHA)

    # Bootstrap CI iid
    ci_iid_lo, ci_iid_hi = bootstrap_sr_iid(dr_arr, alpha=AS_ALPHA)

    # Bootstrap CI block
    ci_blk_lo, ci_blk_hi = bootstrap_sr_block(dr_arr, block_size=block_size,
                                                alpha=AS_ALPHA)

    # DA su target smoothato
    ref_da = np.sign(_sm)
    da     = float(np.mean(ref_da == np.sign(pred)) * 100)

    records.append({
        'Model':        mname,
        'T_active':     T_active,
        'SR_ann':       sr,
        'Skew(γ3)':     skew,
        'Kurt(γ4)':     kurt_p,
        'AR(1)':        float(_stats.pearsonr(inv_r[:-1], inv_r[1:])[0])
                        if T_active > 5 else np.nan,
        'JB_p':         jb_p,
        'LB10_p':       lb_p,
        'PSR(SR*=0)':   psr_0,
        'PSR(SR*=BH)':  psr_bh,
        'DSR(K)':       dsr_v,
        'SR*_adj':      sr_adj,
        'MinTRL(d)':    mtrl,
        'CI_iid_lo':    ci_iid_lo,
        'CI_iid_hi':    ci_iid_hi,
        'CI_blk_lo':    ci_blk_lo,
        'CI_blk_hi':    ci_blk_hi,
        'DA%':          da,
    })

df_as = pd.DataFrame(records).set_index('Model')


# ════════════════════════════════════════════════════════════
# PRINT TABELLE
# ════════════════════════════════════════════════════════════

SEP = '  ' + '─'*95

print('\n' + '='*95)
print(f'  ADVANCED SHARPE ANALYSIS — Strategy A (Long/Flat) | Test OOS')
print(f'  SR*={AS_SR_STAR} | α={AS_ALPHA} | K={AS_K_TRIALS} | block_size={block_size}d')
print(f'  Buy & Hold SR (calendario) = {bh_sr:+.3f}')
print('='*95)

# ── Tabella 1: statistiche descrittive ritorni ───────────────
print(f'\n  {"Modello":<14} {"T_act":>6} {"SR_ann":>8} {"Skew(γ3)":>10} '
      f'{"Kurt(γ4)":>10} {"AR(1)":>7} {"JB_p":>7} {"LB10_p":>8} {"DA%":>6}')
print(SEP)
for _, row in df_as.iterrows():
    mn   = row.name
    norm = '✅' if row['JB_p'] > 0.05  else '❌'
    iid  = '✅' if row['LB10_p'] > 0.05 else '❌'
    sr_s = f'{row["SR_ann"]:+.3f}' if not np.isnan(row['SR_ann']) else '  nan '
    sk_s = f'{row["Skew(γ3)"]:+.3f}' if not np.isnan(row['Skew(γ3)']) else '  nan'
    ku_s = f'{row["Kurt(γ4)"]:+.3f}' if not np.isnan(row['Kurt(γ4)']) else '  nan'
    ar_s = f'{row["AR(1)"]:+.3f}'    if not np.isnan(row['AR(1)'])    else '  nan'
    jb_s = f'{row["JB_p"]:.3f}{norm}' if not np.isnan(row['JB_p'])   else ' nan'
    lb_s = f'{row["LB10_p"]:.3f}{iid}' if not np.isnan(row['LB10_p']) else ' nan'
    print(f'  {mn:<14} {int(row["T_active"]):>6} {sr_s:>8} {sk_s:>10} '
          f'{ku_s:>10} {ar_s:>7} {jb_s:>8} {lb_s:>9} {row["DA%"]:>5.1f}%')

print(f'\n  JB_p > 0.05 ✅ = distribuzione normale | LB10_p > 0.05 ✅ = assenza autocorrelazione')
print(f'  SR calcolato su T_act giorni investiti (esclude giorni flat con ret=0)')

# ── Tabella 2: PSR / DSR / MinTRL ───────────────────────────
print(f'\n  {"Modello":<14} {"PSR(SR*=0)":>12} {"PSR(SR*=BH)":>13} '
      f'{"DSR(K={AS_K_TRIALS})":>12} {"SR*_adj":>9} {"MinTRL(d)":>10}')
print(SEP)
for _, row in df_as.iterrows():
    mn   = row.name
    p0   = f'{row["PSR(SR*=0)"]:.4f}'  if not np.isnan(row['PSR(SR*=0)'])  else '   nan'
    pbh  = f'{row["PSR(SR*=BH)"]:.4f}' if not np.isnan(row['PSR(SR*=BH)']) else '   nan'
    dv   = f'{row["DSR(K)"]:.4f}'      if not np.isnan(row['DSR(K)'])      else '   nan'
    sa   = f'{row["SR*_adj"]:+.3f}'    if not np.isnan(row['SR*_adj'])      else '   nan'
    mt   = f'{row["MinTRL(d)"]:.1f}d'  if not np.isinf(row['MinTRL(d)']) and \
                                            not np.isnan(row['MinTRL(d)'])  else '   ∞'
    sig0 = ' ✅' if not np.isnan(row['PSR(SR*=0)'])  and row['PSR(SR*=0)']  > (1-AS_ALPHA) else ' ⚠️'
    sigd = ' ✅' if not np.isnan(row['DSR(K)'])       and row['DSR(K)']      > (1-AS_ALPHA) else ' ⚠️'
    print(f'  {mn:<14} {p0:>10}{sig0}  {pbh:>11}  {dv:>10}{sigd}  {sa:>9}  {mt:>9}')

print(f'\n  PSR > {1-AS_ALPHA:.2f} ✅ = significativo al {int(AS_ALPHA*100)}%')
print(f'  DSR corregge PSR per K={AS_K_TRIALS} configurazioni testate (60 Optuna + 1 ensemble)')
print(f'  SR*_adj = SR benchmark atteso per caso su K={AS_K_TRIALS} trial (Extreme Value Theory)')
print(f'  MinTRL = giorni investiti minimi per concludere SR>SR*=0 con α={AS_ALPHA}')

# ── Tabella 3: Bootstrap CI ──────────────────────────────────
print(f'\n  {"Modello":<14} {"CI_iid_lo":>10} {"CI_iid_hi":>10} {"Sig_iid":>8}'
      f'  {"CI_blk_lo":>10} {"CI_blk_hi":>10} {"Sig_blk":>8}')
print(SEP)
for _, row in df_as.iterrows():
    mn    = row.name
    def _fmt(v): return f'{v:+.3f}' if not np.isnan(v) else '  nan '
    sig_i = '✅ excl.0' if not np.isnan(row['CI_iid_lo']) and row['CI_iid_lo'] > 0 \
            else '⚠️ incl.0'
    sig_b = '✅ excl.0' if not np.isnan(row['CI_blk_lo']) and row['CI_blk_lo'] > 0 \
            else '⚠️ incl.0'
    print(f'  {mn:<14} {_fmt(row["CI_iid_lo"]):>10} {_fmt(row["CI_iid_hi"]):>10} '
          f'{sig_i:>8}  {_fmt(row["CI_blk_lo"]):>10} {_fmt(row["CI_blk_hi"]):>10} '
          f'{sig_b:>8}')

print(f'\n  Bootstrap iid: ricampiona giorni investiti con replacement (1000 resamples)')
print(f'  Block bootstrap: block_size={block_size}d (≈√T={np.sqrt(T_cal):.1f}) — preserva autocorrelazione')
print(f'  CI esclude 0 ✅ = SR positivo con confidenza {int((1-AS_ALPHA)*100)}%')


# ════════════════════════════════════════════════════════════
# GRAFICI — distribuzione SR bootstrap per ogni modello
# ════════════════════════════════════════════════════════════

n_mod  = len(AS_MODELS_OK)
n_cols = min(4, n_mod)
n_rows = int(np.ceil(n_mod / n_cols))
fig, axes = plt.subplots(n_rows, n_cols,
                          figsize=(n_cols * 5, n_rows * 3.5))
axes_flat = np.array(axes).flatten() if n_mod > 1 else [axes]
fig.suptitle(
    f'Bootstrap SR Distribution — Strategy A | '
    f'SR*={AS_SR_STAR} | block_size={block_size}d | K={AS_K_TRIALS}',
    fontsize=12, fontweight='bold')

for i, mname in enumerate(AS_MODELS_OK):
    ax   = axes_flat[i]
    row  = df_as.loc[mname]
    pred = all_preds_test[mname]

    _take = TS_TAKE_PR if 'TS_TAKE_PR' in dir() else 0.0005
    _stop = TS_STOP_L  if 'TS_STOP_L'  in dir() else -0.0005
    port  = 100_000.0; pos = 0; dr = []
    for t in range(T_cal):
        ret_t = np.exp(lr_test_true[t]) - 1
        if pred[t] > _take:   new_pos = 1
        elif pred[t] < _stop: new_pos = 0
        else:                  new_pos = pos
        tc = 0.001 if new_pos != pos else 0.0
        new_port = port * (1-tc) * (1+ret_t) if new_pos==1 else port*(1-tc)
        dr.append((new_port - port)/port)
        port = new_port; pos = new_pos
    dr_arr = np.array(dr)

    # Ricalcola bootstrap per il grafico
    inv_r  = dr_arr[dr_arr != 0]
    np.random.seed(42)
    boot_i = np.array([
        (lambda x: (x.mean()/(x.std(ddof=1)+1e-12))*np.sqrt(252))(
            np.random.choice(inv_r, len(inv_r), replace=True))
        for _ in range(1000)
    ])
    T_a    = len(inv_r)
    bs2    = max(2, int(np.floor(np.sqrt(T_a))))
    np.random.seed(43)
    boot_b = []
    if T_a >= 5 and T_a > bs2:
        for _ in range(1000):
            idx = []
            while len(idx) < T_a:
                s = np.random.randint(0, max(1, T_a - bs2 + 1))
                idx.extend(list(range(s, min(s + bs2, T_a))))
            idx_arr = np.array(idx[:T_a], dtype=np.intp)
            boot_b.append((inv_r[idx_arr].mean() /
                           (inv_r[idx_arr].std(ddof=1) + 1e-12)) * np.sqrt(252))
    boot_b = np.array(boot_b) if boot_b else np.array([np.nan])

    ci_i   = np.percentile(boot_i, [2.5, 97.5])
    ci_b   = np.percentile(boot_b, [2.5, 97.5])
    sr_obs = row['SR_ann']

    # Grafico
    ax.hist(boot_i, bins=40, alpha=0.5, color='steelblue', label='iid', density=True)
    ax.hist(boot_b, bins=40, alpha=0.5, color='#e67e22',   label=f'block({bs2}d)', density=True)
    ax.axvline(sr_obs,  color='black',  lw=2,   ls='-',  label=f'SR_obs={sr_obs:+.2f}')
    ax.axvline(AS_SR_STAR, color='red', lw=1.5, ls='--', label=f'SR*={AS_SR_STAR}')
    ax.axvline(ci_i[0], color='steelblue', lw=1, ls=':', alpha=0.8)
    ax.axvline(ci_i[1], color='steelblue', lw=1, ls=':', alpha=0.8)
    ax.axvline(ci_b[0], color='#e67e22',   lw=1, ls=':', alpha=0.8)
    ax.axvline(ci_b[1], color='#e67e22',   lw=1, ls=':', alpha=0.8)

    psr0_str = f'{row["PSR(SR*=0)"]:.3f}' if not np.isnan(row['PSR(SR*=0)']) else 'nan'
    dsr_str  = f'{row["DSR(K)"]:.3f}'     if not np.isnan(row['DSR(K)'])     else 'nan'
    ax.set_title(f'{mname}\nSR={sr_obs:+.2f} | PSR={psr0_str} | DSR={dsr_str}',
                 fontsize=9, fontweight='bold')
    ax.legend(fontsize=7); ax.grid(alpha=0.3)
    ax.set_xlabel('Annualised Sharpe Ratio', fontsize=8)

for j in range(n_mod, len(axes_flat)):
    axes_flat[j].set_visible(False)

plt.tight_layout(); plt.show()


# ════════════════════════════════════════════════════════════
# NOTE INTERPRETATIVE
# ════════════════════════════════════════════════════════════

print('\n' + '='*70)
print('  NOTE INTERPRETATIVE')
print('='*70)
print(f'  SR calcolato su T_active (giorni investiti, flat esclusi) per evitare')
print(f'  inflazione da giorni a rendimento zero che comprimono σ artificialmente.')
print()
print(f'  PSR(SR*=0) = probabilità che il vero SR > 0, corretta per γ3 e γ4.')
print(f'  PSR(SR*=BH) = probabilità che la strategia batta il Buy & Hold.')
print()
print(f'  DSR(K={AS_K_TRIALS}) corregge per selezione su K configurazioni:')
print(f'    SR*_adj = massimo SR atteso per caso su {AS_K_TRIALS} trial (EVT).')
print(f'    DSR > {1-AS_ALPHA:.2f} ✅ = il risultato sopravvive alla correzione multipla al {int(AS_ALPHA*100)}%.')
print()
print(f'  MinTRL = giorni investiti minimi per concludere SR>0 con α={AS_ALPHA}.')
print(f'    MinTRL << T_active → track record sufficiente.')
print(f'    MinTRL >> T_active → track record insufficiente per questa stima SR.')
print()
print(f'  Block bootstrap (block_size={block_size}d) preferibile quando LB10_p < 0.05')
print(f'  (autocorrelazione presente) — iid sottostima la varianza in quel caso.')
print('='*70)
```
